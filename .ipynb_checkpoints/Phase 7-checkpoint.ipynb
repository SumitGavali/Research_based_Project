```python
# Cell 2.6: Create Predictions from Model (FIXED)
print("\n" + "="*60)
print("CREATING PREDICTIONS FROM MODEL")
print("="*60)

import glob
import joblib
import pandas as pd
import numpy as np

# 1. Load model
model_files = glob.glob('final_rf_model_*.pkl')
if not model_files:
    print("❌ Model not found!")
else:
    rf_model = joblib.load(sorted(model_files)[-1])
    print(f"✅ Loaded Random Forest model: {sorted(model_files)[-1]}")
    
    # 2. Load feature list (the top features used in training)
    feature_files = glob.glob('final_features_*.csv')
    if feature_files:
        feature_list = pd.read_csv(sorted(feature_files)[-1])
        top_features = feature_list['Feature'].tolist()
        print(f"✅ Loaded feature list: {len(top_features)} features")
    else:
        # Try to get feature names from model
        try:
            top_features = rf_model.feature_names_in_
            print(f"✅ Got features from model: {len(top_features)} features")
        except:
            print("❌ Could not get feature list")
            top_features = None
    
    # 3. Load anchored features
    anchored_files = glob.glob('anchored_features_*.csv')
    if anchored_files:
        X = pd.read_csv(sorted(anchored_files)[-1], index_col=0, parse_dates=True)
        print(f"✅ Loaded anchored features: {sorted(anchored_files)[-1]}")
        print(f"   Shape: {X.shape}")
    else:
        # Load regular features and create anchors
        feature_files = glob.glob('features_X_*.csv')
        X = pd.read_csv(sorted(feature_files)[-1], index_col=0, parse_dates=True)
        print(f"✅ Loaded features: {sorted(feature_files)[-1]}")
        
        # Add anchor features
        spy_returns = master_data['SPY'].pct_change()
        X['vol_floor_50'] = spy_returns.rolling(50).std() * np.sqrt(252)
        X['vol_floor_200'] = spy_returns.rolling(200).std() * np.sqrt(252)
        X = X.ffill().bfill().fillna(0)
        print(f"✅ Added anchor features")
    
    # 4. Select only the features the model expects
    if top_features is not None:
        # Check which features are available
        available_features = [f for f in top_features if f in X.columns]
        missing_features = [f for f in top_features if f not in X.columns]
        
        if missing_features:
            print(f"⚠️ Missing {len(missing_features)} features, creating them...")
            for feat in missing_features:
                if 'floor' in feat:
                    # Already handled above
                    pass
                else:
                    X[feat] = 0
            print(f"✅ Added missing features with zeros")
        
        # Select only top features in correct order
        X_selected = X[top_features].copy()
        print(f"✅ Selected {len(top_features)} features for prediction")
    else:
        # Use all features
        X_selected = X.copy()
        print(f"⚠️ Using all {X_selected.shape[1]} features")
    
    # 5. Align with y_test
    common_idx = X_selected.index.intersection(y_test.index)
    X_aligned = X_selected.loc[common_idx]
    y_test_aligned = y_test.loc[common_idx]
    
    print(f"✅ Aligned data: {len(X_aligned)} observations")
    
    # 6. Make predictions
    rf_pred_imp = rf_model.predict(X_aligned)
    print(f"✅ Generated predictions for {len(rf_pred_imp)} observations")
    
    # 7. Store in global scope
    globals()['rf_pred_imp'] = rf_pred_imp
    globals()['y_test'] = y_test_aligned
    
    print(f"\n✅ rf_pred_imp shape: {rf_pred_imp.shape}")
    print(f"✅ y_test shape: {y_test_aligned.shape}")
    print(f"✅ First 5 predictions: {rf_pred_imp[:5]}")
```

```python
# Cell 3: Prepare Backtest Data (RUN NOW - should work!)
def prepare_backtest_data(master_data, y_test, rf_pred_imp):
    """
    Prepare data for backtesting with proper alignment
    """
    print("\n" + "="*60)
    print("PREPARING BACKTEST DATA")
    print("="*60)
    
    # Get SPY prices and returns
    spy_prices = master_data['SPY']
    spy_returns = spy_prices.pct_change()
    
    # Get test period
    test_start = y_test.index[0]
    test_end = y_test.index[-1]
    
    print(f"Test Period: {test_start} to {test_end}")
    
    # Get SPY data for test period
    spy_prices_test = spy_prices.loc[test_start:test_end]
    spy_returns_test = spy_returns.loc[test_start:test_end]
    
    # Get predictions (in log space) and convert back to volatility
    if hasattr(rf_pred_imp, 'values'):
        predictions_log = rf_pred_imp.values.flatten()
    else:
        predictions_log = rf_pred_imp.flatten()
    
    # Convert from log space back to volatility
    predictions = np.exp(predictions_log)
    
    # Ensure same length
    min_len = min(len(spy_prices_test), len(predictions))
    
    # Create DataFrame with proper alignment
    backtest_data = pd.DataFrame({
        'SPY_Price': spy_prices_test.iloc[:min_len],
        'SPY_Return': spy_returns_test.iloc[:min_len],
        'Predicted_Vol': predictions[:min_len]
    }, index=spy_prices_test.index[:min_len])
    
    # Remove NaN values
    backtest_data = backtest_data.dropna()
    
    print(f"\nBacktest Data Shape: {backtest_data.shape}")
    print(f"Date Range: {backtest_data.index[0]} to {backtest_data.index[-1]}")
    print(f"Trading Days: {len(backtest_data)}")
    
    # Check for any remaining issues
    print(f"\nData Statistics:")
    print(f"  Mean Predicted Vol: {backtest_data['Predicted_Vol'].mean():.4f}")
    print(f"  Min Predicted Vol: {backtest_data['Predicted_Vol'].min():.4f}")
    print(f"  Max Predicted Vol: {backtest_data['Predicted_Vol'].max():.4f}")
    print(f"  Mean SPY Return: {backtest_data['SPY_Return'].mean():.6f}")
    
    return backtest_data

# Prepare backtest data
backtest_data = prepare_backtest_data(master_data, y_test, rf_pred_imp)
```

```python
# Cell 4: Implement Volatility Targeting Strategy
def implement_strategy(backtest_data, target_vol=0.15, min_position=0.3, max_position=2.0):
    """
    Implement volatility targeting strategy
    """
    print("\n" + "="*60)
    print("IMPLEMENTING VOLATILITY TARGETING STRATEGY")
    print("="*60)
    
    # Position sizing (inverse volatility)
    position_size = target_vol / backtest_data['Predicted_Vol']
    
    # Clip position sizes
    position_size = np.clip(position_size, min_position, max_position)
    
    # Calculate strategy returns (using previous day's position)
    strategy_returns = backtest_data['SPY_Return'] * position_size.shift(1)
    
    # Calculate cumulative returns
    strategy_cumulative = (1 + strategy_returns).cumprod()
    buy_hold_cumulative = (1 + backtest_data['SPY_Return']).cumprod()
    
    # Add to backtest data
    backtest_data['Position_Size'] = position_size
    backtest_data['Strategy_Return'] = strategy_returns
    backtest_data['Strategy_Cumulative'] = strategy_cumulative
    backtest_data['Buy_Hold_Cumulative'] = buy_hold_cumulative
    
    print(f"\nStrategy Statistics:")
    print(f"  Average Position Size: {position_size.mean():.2f}")
    print(f"  Min Position Size: {position_size.min():.2f}")
    print(f"  Max Position Size: {position_size.max():.2f}")
    print(f"  Volatility Target: {target_vol:.1%}")
    
    return backtest_data

# Implement strategy
backtest_data = implement_strategy(backtest_data)
```

```python
# Cell 5: Calculate Performance Metrics
def calculate_performance_metrics(backtest_data):
    """
    Calculate comprehensive performance metrics
    """
    print("\n" + "="*60)
    print("CALCULATING PERFORMANCE METRICS")
    print("="*60)
    
    # Extract returns
    strategy_returns = backtest_data['Strategy_Return'].dropna()
    buy_hold_returns = backtest_data['SPY_Return'].dropna()
    
    # Strategy metrics
    strategy_metrics = {
        'Total_Return': (backtest_data['Strategy_Cumulative'].iloc[-1] - 1),
        'Annualized_Return': (backtest_data['Strategy_Cumulative'].iloc[-1]) ** (252 / len(strategy_returns)) - 1,
        'Annualized_Volatility': strategy_returns.std() * np.sqrt(252),
        'Sharpe_Ratio': (strategy_returns.mean() / strategy_returns.std()) * np.sqrt(252) if strategy_returns.std() > 0 else 0,
        'Max_Drawdown': (backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min(),
        'Win_Rate': (strategy_returns > 0).sum() / len(strategy_returns),
        'Avg_Win': strategy_returns[strategy_returns > 0].mean() if (strategy_returns > 0).sum() > 0 else 0,
        'Avg_Loss': strategy_returns[strategy_returns < 0].mean() if (strategy_returns < 0).sum() > 0 else 0,
        'Profit_Factor': abs(strategy_returns[strategy_returns > 0].sum() / strategy_returns[strategy_returns < 0].sum()) if (strategy_returns < 0).sum() > 0 else 0,
        'Calmar_Ratio': ((backtest_data['Strategy_Cumulative'].iloc[-1] - 1) / abs((backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min())) if (backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min() != 0 else 0
    }
    
    # Buy & Hold metrics
    bh_metrics = {
        'Total_Return': (backtest_data['Buy_Hold_Cumulative'].iloc[-1] - 1),
        'Annualized_Return': (backtest_data['Buy_Hold_Cumulative'].iloc[-1]) ** (252 / len(buy_hold_returns)) - 1,
        'Annualized_Volatility': buy_hold_returns.std() * np.sqrt(252),
        'Sharpe_Ratio': (buy_hold_returns.mean() / buy_hold_returns.std()) * np.sqrt(252) if buy_hold_returns.std() > 0 else 0,
        'Max_Drawdown': (backtest_data['Buy_Hold_Cumulative'] / backtest_data['Buy_Hold_Cumulative'].cummax() - 1).min(),
        'Win_Rate': (buy_hold_returns > 0).sum() / len(buy_hold_returns)
    }
    
    # Print results
    print("\n📊 Strategy Performance:")
    print("-"*50)
    for key, value in strategy_metrics.items():
        if 'Return' in key or 'Volatility' in key:
            print(f"  {key.replace('_', ' ')}: {value:.2%}")
        else:
            print(f"  {key.replace('_', ' ')}: {value:.4f}")
    
    print("\n📊 Buy & Hold Performance:")
    print("-"*50)
    for key, value in bh_metrics.items():
        if 'Return' in key or 'Volatility' in key:
            print(f"  {key.replace('_', ' ')}: {value:.2%}")
        else:
            print(f"  {key.replace('_', ' ')}: {value:.4f}")
    
    # Calculate outperformance
    outperformance = {
        'Excess_Return': strategy_metrics['Total_Return'] - bh_metrics['Total_Return'],
        'Sharpe_Improvement': strategy_metrics['Sharpe_Ratio'] - bh_metrics['Sharpe_Ratio'],
        'Drawdown_Reduction': bh_metrics['Max_Drawdown'] - strategy_metrics['Max_Drawdown']
    }
    
    print("\n📈 Strategy vs Buy & Hold:")
    print("-"*50)
    print(f"  Excess Return: {outperformance['Excess_Return']:.2%}")
    print(f"  Sharpe Improvement: {outperformance['Sharpe_Improvement']:.4f}")
    print(f"  Drawdown Reduction: {outperformance['Drawdown_Reduction']:.2%}")
    
    return strategy_metrics, bh_metrics, outperformance

# Calculate metrics
strategy_metrics, bh_metrics, outperformance = calculate_performance_metrics(backtest_data)
```

Buy & Hold represents the passive market benchmark (S&P 500).\(just invest not analysis before)\
Strategy Performance represents our active Machine Learning system. The backtest proves that our \
Random Forest model successfully provided a mathematical edge, delivering an Excess Return of 97.36% over the passive market baseline while maintaining a safer drawdown profile.

```python
# Cell 7: Monthly Performance Analysis
def analyze_monthly_performance(backtest_data):
    """
    Analyze performance by month
    """
    print("\n" + "="*60)
    print("MONTHLY PERFORMANCE ANALYSIS")
    print("="*60)
    
    # Create monthly returns
    monthly_returns = backtest_data['Strategy_Return'].resample('ME').apply(lambda x: (1 + x).prod() - 1)
    monthly_bh = backtest_data['SPY_Return'].resample('ME').apply(lambda x: (1 + x).prod() - 1)
    
    # Create monthly performance table
    monthly_df = pd.DataFrame({
        'Strategy': monthly_returns,
        'Buy_Hold': monthly_bh,
        'Excess': monthly_returns - monthly_bh
    })
    
    print("\nMonthly Performance Summary:")
    print("-"*50)
    print(f"  Strategy Mean Monthly Return: {monthly_returns.mean():.2%}")
    print(f"  Strategy Std Monthly Return: {monthly_returns.std():.2%}")
    print(f"  Strategy Win Rate: {(monthly_returns > 0).sum() / len(monthly_returns):.1%}")
    print(f"  Buy & Hold Win Rate: {(monthly_bh > 0).sum() / len(monthly_bh):.1%}")
    print(f"  Strategy vs BH Win Rate: {(monthly_df['Excess'] > 0).sum() / len(monthly_df):.1%}")
    
    # Best and worst months
    print("\nBest 5 Months (Strategy):")
    print("-"*50)
    for date, ret in monthly_returns.nlargest(5).items():
        print(f"  {date.strftime('%Y-%m')}: {ret:.2%}")
    
    print("\nWorst 5 Months (Strategy):")
    print("-"*50)
    for date, ret in monthly_returns.nsmallest(5).items():
        print(f"  {date.strftime('%Y-%m')}: {ret:.2%}")
    
    # Heatmap of monthly returns
    monthly_pivot = monthly_returns.groupby([monthly_returns.index.year, monthly_returns.index.month]).mean().unstack()
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Strategy monthly returns
    ax1 = axes[0]
    sns.heatmap(monthly_pivot, annot=True, fmt='.2%', cmap='RdYlGn', center=0, ax=ax1)
    ax1.set_title('Strategy Monthly Returns (%)')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Year')
    
    # Excess returns vs Buy & Hold
    excess_pivot = monthly_df['Excess'].groupby([monthly_df.index.year, monthly_df.index.month]).mean().unstack()
    ax2 = axes[1]
    sns.heatmap(excess_pivot, annot=True, fmt='.2%', cmap='RdYlGn', center=0, ax=ax2)
    ax2.set_title('Excess Returns vs Buy & Hold (%)')
    ax2.set_xlabel('Month')
    ax2.set_ylabel('Year')
    
    plt.tight_layout()
    plt.show()
    
    return monthly_df

# Analyze monthly performance
monthly_performance = analyze_monthly_performance(backtest_data)
```

```python
# Cell 8: Risk-Adjusted Metrics
def calculate_risk_adjusted_metrics(backtest_data):
    """
    Calculate advanced risk-adjusted metrics
    """
    print("\n" + "="*60)
    print("RISK-ADJUSTED METRICS")
    print("="*60)
    
    returns = backtest_data['Strategy_Return'].dropna()
    
    # Calculate additional metrics
    metrics = {
        'Sortino_Ratio': (returns.mean() / returns[returns < 0].std()) * np.sqrt(252) if len(returns[returns < 0]) > 0 else 0,
        'Calmar_Ratio': (backtest_data['Strategy_Cumulative'].iloc[-1] - 1) / abs(backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min() if (backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min() != 0 else 0,
        'Max_Drawdown': (backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1).min(),
        'Max_Drawdown_Duration': (backtest_data['Strategy_Cumulative'] / backtest_data['Strategy_Cumulative'].cummax() - 1 < 0).sum(),
        'Up_Capture': (returns[returns > 0].sum() / backtest_data['SPY_Return'][backtest_data['SPY_Return'] > 0].sum()) if (backtest_data['SPY_Return'] > 0).sum() > 0 else 0,
        'Down_Capture': (returns[returns < 0].sum() / backtest_data['SPY_Return'][backtest_data['SPY_Return'] < 0].sum()) if (backtest_data['SPY_Return'] < 0).sum() > 0 else 0,
        'Capture_Ratio': None  # Will calculate below
    }
    
    metrics['Capture_Ratio'] = metrics['Up_Capture'] / metrics['Down_Capture'] if metrics['Down_Capture'] != 0 else 0
    
    print("\nRisk-Adjusted Metrics:")
    print("-"*50)
    print(f"  Sortino Ratio: {metrics['Sortino_Ratio']:.4f}")
    print(f"  Calmar Ratio: {metrics['Calmar_Ratio']:.4f}")
    print(f"  Max Drawdown: {metrics['Max_Drawdown']:.2%}")
    print(f"  Max Drawdown Duration: {metrics['Max_Drawdown_Duration']} days")
    print(f"  Up Capture: {metrics['Up_Capture']:.2%}")
    print(f"  Down Capture: {metrics['Down_Capture']:.2%}")
    print(f"  Capture Ratio: {metrics['Capture_Ratio']:.2f}")
    
    return metrics

# Calculate risk-adjusted metrics
risk_metrics = calculate_risk_adjusted_metrics(backtest_data)
```

#### Instead of just looking at raw profits, they measure how much psychological "pain" and risk your system endured to achieve that 770% return over 20 years.

 *Sortino Ratio*: Unlike the Sharpe ratio, which penalizes a model even if it makes a massive sudden profit, the Sortino ratio completely **ignores positive** upside spikes and only punishes the model for actual drops.

 *Max draw down duration* :This tracks the timeline of your worst investment valley. It counts the number of days it took for the portfolio to fall from a historic peak, hit the bottom of the 2008 financial crisis, and climb all the way back up to break even\
**Insight**: 4,333 days is roughly 11.8 years. This highlights the severe reality of backtesting over the 2008 crash. Because the global economy collapsed so heavily, a simple passive investor (and your ML strategy, which tracked it closely) spent over a decade just fighting to recover its lost wealth before printing new net profits.



```python
# Cell 9: Save Backtest Results
def save_backtest_results(backtest_data, strategy_metrics, bh_metrics, outperformance, risk_metrics):
    """
    Save all backtest results
    """
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    
    print("\n" + "="*60)
    print("SAVING BACKTEST RESULTS")
    print("="*60)
    
    # Save backtest data
    backtest_data.to_csv(f'backtest_results_{timestamp}.csv')
    print(f"✅ Backtest data saved: backtest_results_{timestamp}.csv")
    
    # Save strategy metrics
    metrics_df = pd.DataFrame({
        'Metric': list(strategy_metrics.keys()) + list(bh_metrics.keys()),
        'Strategy': list(strategy_metrics.values()) + [''] * len(bh_metrics.keys()),
        'Buy_Hold': [''] * len(strategy_metrics.keys()) + list(bh_metrics.values())
    })
    metrics_df.to_csv(f'performance_metrics_{timestamp}.csv', index=False)
    print(f"✅ Performance metrics saved: performance_metrics_{timestamp}.csv")
    
    # Save outperformance
    outperformance_df = pd.DataFrame({
        'Metric': list(outperformance.keys()),
        'Value': list(outperformance.values())
    })
    outperformance_df.to_csv(f'outperformance_{timestamp}.csv', index=False)
    print(f"✅ Outperformance saved: outperformance_{timestamp}.csv")
    
    # Save risk metrics
    risk_df = pd.DataFrame({
        'Metric': list(risk_metrics.keys()),
        'Value': list(risk_metrics.values())
    })
    risk_df.to_csv(f'risk_metrics_{timestamp}.csv', index=False)
    print(f"✅ Risk metrics saved: risk_metrics_{timestamp}.csv")
    
    # Save monthly performance
    if 'monthly_performance' in globals():
        monthly_performance.to_csv(f'monthly_performance_{timestamp}.csv')
        print(f"✅ Monthly performance saved: monthly_performance_{timestamp}.csv")
    
    print("\n✅ All backtest results saved successfully!")
    
    return timestamp

# Save all results
backtest_timestamp = save_backtest_results(
    backtest_data, strategy_metrics, bh_metrics, 
    outperformance, risk_metrics
)
```
