### Feature engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')


pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 20)
```

```python
# Cell 2: Load Data from Previous Phases
def load_data():
    """Load master data and returns data from previous phases"""
    import glob
    
    # Load master data
    master_files = glob.glob('master_data_*.csv')
    if master_files:
        latest_master = sorted(master_files)[-1]
        master_data = pd.read_csv(latest_master, index_col=0, parse_dates=True)
        print(f"loaded master data: {latest_master}")
    else:
        print(" No master data found!")
        return None, None
    
    # Load returns data
    returns_files = glob.glob('returns_data_*.csv')
    if returns_files:
        latest_returns = sorted(returns_files)[-1]
        returns_data = pd.read_csv(latest_returns, index_col=0, parse_dates=True)
        print(f" Loaded returns data: {latest_returns}")
    else:
        print(" No returns data found, creating from master...")
        returns_data = pd.DataFrame()
        for col in master_data.columns:
            returns_data[f'{col}_ret'] = master_data[col].pct_change()
            returns_data[f'{col}_logret'] = np.log(master_data[col] / master_data[col].shift(1))
    
    print(f" Master shape: {master_data.shape}")
    print(f" Returns shape: {returns_data.shape}")
    print(f" Date range: {master_data.index[0]} to {master_data.index[-1]}")
    
    return master_data, returns_data

master_data, returns_data = load_data()
```

## Lag feature 

```python
# Cell 3: Lag Features
def create_lag_features(data, asset, lags=[1, 2, 3, 5, 10, 20]):
    """
    Create lag features for a given asset
    """
    print(f"\nCreating lag features for {asset}...")
    
    feature_df = pd.DataFrame(index=data.index)
    
    # Lags of returns
    ret_col = f'{asset}_ret'
    if ret_col in data.columns:
        for lag in lags:
            feature_df[f'{asset}_lag{lag}'] = data[ret_col].shift(lag)
            print(f"   {asset}_lag{lag}")
    
    # Lags of log returns
    logret_col = f'{asset}_logret'
    if logret_col in data.columns:
        for lag in lags[:5]:  # Fewer lags for log returns
            feature_df[f'{asset}_loglag{lag}'] = data[logret_col].shift(lag)
    
    # Lags of squared returns (volatility proxy)
    if ret_col in data.columns:
        squared_ret = data[ret_col] ** 2
        for lag in lags[:5]:
            feature_df[f'{asset}_sq_lag{lag}'] = squared_ret.shift(lag)
    
    print(f"Created {len(feature_df.columns)} lag features")
    return feature_df

# Create lag features for SPY
spy_lags = create_lag_features(returns_data, 'SPY')
print(f"\nSPY Lag Features Preview:")
print(spy_lags.head())
```

We built an automated data recovery and feature-engineering pipeline. To prepare our dataset for predictive time-series modeling, we converted continuous returns into discrete Lag Features across 1, 2, 3, 5, 10, and 20-day historical intervals using Pandas .shift(). Furthermore, we calculated squared return lags as an internal proxy for time-varying historical volatility. This structural alignment shifts historical data points from previous trading periods directly into the current row observation plane. This removes forward-looking data leakage and provides our machine learning models with a clean, structured time-series history to find patterns between past price turbulence and tomorrow's expected returns.

### Rolling Statistics Features

```python
# Cell 4: Rolling Statistics Features
def create_rolling_features(data, asset, windows=[5, 10, 20, 50, 100]):
    
    print(f"\nCreating rolling features for {asset}...")
    
    feature_df = pd.DataFrame(index=data.index)
    ret_col = f'{asset}_ret'
    
    if ret_col not in data.columns:
        print(f" {ret_col} not found!")
        return feature_df
    
    returns = data[ret_col]
    
    for window in windows:
        # Rolling mean (trend)
        feature_df[f'{asset}_roll_mean_{window}'] = returns.rolling(window).mean()
        
        # Rolling std (volatility)
        feature_df[f'{asset}_roll_std_{window}'] = returns.rolling(window).std()
        
        # Rolling skewness (asymmetry)
        feature_df[f'{asset}_roll_skew_{window}'] = returns.rolling(window).skew()
        
        # Rolling kurtosis (tail risk)
        feature_df[f'{asset}_roll_kurt_{window}'] = returns.rolling(window).kurt()
        
        # Rolling min and max (range)
        feature_df[f'{asset}_roll_min_{window}'] = returns.rolling(window).min()
        feature_df[f'{asset}_roll_max_{window}'] = returns.rolling(window).max()
        
        # Rolling quantiles (risk measures)
        feature_df[f'{asset}_roll_q25_{window}'] = returns.rolling(window).quantile(0.25)
        feature_df[f'{asset}_roll_q75_{window}'] = returns.rolling(window).quantile(0.75)
        
        print(f"   Window {window} completed")
    
    print(f" Created {len(feature_df.columns)} rolling features")
    return feature_df

# Create rolling features for SPY
spy_rolling = create_rolling_features(returns_data, 'SPY')
print(f"\nSPY Rolling Features Preview:")
print(spy_rolling.head())
```

### rolling statistics features
_roll_mean (Rolling Trend):\
What: The average percentage return over the last X days.\
Why: Tells the AI if the market is currently riding an upward or downward wave.

_roll_std (Rolling Volatility):\
What: The standard deviation of recent returns.\
Why: Feeds your "yesterday high, today high" volatility clustering logic directly into the model.

_roll_skew & _roll_kurt (Shape Trackers):\
What: Measures the shifting balance and thickness of the data tails.\
Why: Warns the AI if the risk profile is bending toward a sudden "fat-tailed" crash event.

_roll_q25 & _roll_q75 (Percentile Boundaries):\
What: Tracks the threshold of everyday returns.\
Why: Helps the computer isolate normal daily noise from unusual market jumps.

### Technical Indicators


```python
# Cell 5: Technical Indicators
def create_technical_indicators(data, asset):
    """
    Create technical indicator features
    """
    print(f"\nCreating technical indicators for {asset}...")
    
    feature_df = pd.DataFrame(index=data.index)
    prices = master_data[asset]
    ret_col = f'{asset}_ret'
    
    if ret_col not in data.columns:
        print(f"❌ {ret_col} not found!")
        return feature_df
    
    returns = data[ret_col]
    
    # 1. RSI (Relative Strength Index)
    def calculate_rsi(series, period=14):
        """Calculate RSI"""
        delta = series.diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
        rs = gain / loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    feature_df[f'{asset}_RSI_14'] = calculate_rsi(prices, 14)
    feature_df[f'{asset}_RSI_21'] = calculate_rsi(prices, 21)
    print(f"  ✅ RSI features created")
    
    # 2. MACD (Moving Average Convergence Divergence)
    def calculate_macd(series, fast=12, slow=26, signal=9):
        """Calculate MACD"""
        ema_fast = series.ewm(span=fast, adjust=False).mean()
        ema_slow = series.ewm(span=slow, adjust=False).mean()
        macd = ema_fast - ema_slow
        macd_signal = macd.ewm(span=signal, adjust=False).mean()
        macd_hist = macd - macd_signal
        return macd, macd_signal, macd_hist
    
    macd, macd_signal, macd_hist = calculate_macd(prices)
    feature_df[f'{asset}_MACD'] = macd
    feature_df[f'{asset}_MACD_signal'] = macd_signal
    feature_df[f'{asset}_MACD_hist'] = macd_hist
    print(f"  ✅ MACD features created")
    
    # 3. Bollinger Bands
    def calculate_bollinger_bands(series, window=20, num_std=2):
        """Calculate Bollinger Bands"""
        rolling_mean = series.rolling(window).mean()
        rolling_std = series.rolling(window).std()
        upper_band = rolling_mean + (rolling_std * num_std)
        lower_band = rolling_mean - (rolling_std * num_std)
        return rolling_mean, upper_band, lower_band
    
    bb_mid, bb_upper, bb_lower = calculate_bollinger_bands(prices)
    feature_df[f'{asset}_BB_mid'] = bb_mid
    feature_df[f'{asset}_BB_upper'] = bb_upper
    feature_df[f'{asset}_BB_lower'] = bb_lower
    feature_df[f'{asset}_BB_width'] = bb_upper - bb_lower
    feature_df[f'{asset}_BB_position'] = (prices - bb_lower) / (bb_upper - bb_lower)
    print(f"  ✅ Bollinger Bands created")
    
    # 4. Moving Averages
    for window in [5, 10, 20, 50, 200]:
        feature_df[f'{asset}_MA_{window}'] = prices.rolling(window).mean()
        feature_df[f'{asset}_EMA_{window}'] = prices.ewm(span=window, adjust=False).mean()
    
    # Ratio to moving averages
    for window in [20, 50, 200]:
        feature_df[f'{asset}_price_MA{window}_ratio'] = prices / feature_df[f'{asset}_MA_{window}']
    
    print(f"  ✅ Moving averages created")
    
    # 5. Momentum Indicators
    feature_df[f'{asset}_ROC_10'] = prices.pct_change(10) * 100  # Rate of Change
    feature_df[f'{asset}_ROC_20'] = prices.pct_change(20) * 100
    feature_df[f'{asset}_momentum_10'] = prices - prices.shift(10)
    feature_df[f'{asset}_momentum_20'] = prices - prices.shift(20)
    print(f"  ✅ Momentum features created")
    
    # 6. Volatility Indicators
    feature_df[f'{asset}_ATR'] = calculate_atr(master_data, asset)
    print(f"  ✅ ATR created")
    
    print(f"✅ Created {len(feature_df.columns)} technical indicators")
    return feature_df

def calculate_atr(data, asset, period=14):
    """Calculate Average True Range"""
    high = data[asset]  # Using close as proxy for high/low
    low = data[asset]
    close = data[asset].shift(1)
    
    tr1 = high - low
    tr2 = np.abs(high - close)
    tr3 = np.abs(low - close)
    
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    atr = tr.rolling(period).mean()
    return atr

# Create technical indicators for SPY
spy_tech = create_technical_indicators(returns_data, 'SPY')
print(f"\nSPY Technical Indicators Preview:")
print(spy_tech.head())
```

### Technical Indicators (The Trader's Eye)
#### RSI(Relative Strength Index) 
What & Why: A momentum oscillator that scores price speed on a scale from 0 to 100.\
The Use: If RSI crosses above 70, the stock is overbought (too expensive; expect a drop). If it sinks below 30, it is oversold (cheap; expect a rally).

#### MACD (Moving Average Convergence Divergence)
What & Why: Tracks the distance between two exponential moving averages.\
The Use: When the MACD line crosses above its signal line, it proves that short-term buying speed is accelerating. This triggers a "Buy Signal" for  algorithm.

#### Bollinger Bands
What & Why: A 3-line channel drawn around the stock price. The middle line is a moving average, and the upper/lower lines are padded by 2 standard deviations.\
The Use: The bands expand when the market goes wild and shrink when it's calm. _BB_position tells the AI exactly where the current price sits inside the risk channel (0 = touching the bottom floor, 1 = breaking through the ceiling).

#### Moving Averages (MA & EMA)
What & Why: Lines that smooth out daily price noise. EMA gives more weight to recent days, making it react faster.\
The Use: The _price_MA200_ratio tells your model how far the current price has drifted from its long-term 200-day baseline. This is highly useful for identifying massive macro market cycles.

#### Momentum Indicators (ROC)
What & Why: Rate of Change calculates the raw percentage velocity of the stock price over 10 or 20 days.\
The Use: It answers the question: "Is the current price trend speeding up or slowing down?"

#### ATR (Average True Range)
What & Why: Measures the absolute maximum daily price span, tracking trailing market volatility.\
The Use: Algorithms look at the ATR to determine where to place their Stop-Loss safety exits. If ATR is high, the system widens its safety buffer to prevent getting kicked out by random noise.

```python
# Cell 6: Volatility Features
def create_volatility_features(returns_data, asset, windows=[5, 10, 20, 60]):
    """
    Create advanced volatility features
    """
    print(f"\nCreating volatility features for {asset}...")
    
    feature_df = pd.DataFrame(index=returns_data.index)
    ret_col = f'{asset}_ret'
    
    if ret_col not in returns_data.columns:
        print(f"❌ {ret_col} not found!")
        return feature_df
    
    returns = returns_data[ret_col]
    
    # 1. Historical volatility (various windows)
    for window in windows:
        feature_df[f'{asset}_hist_vol_{window}'] = returns.rolling(window).std() * np.sqrt(252)
        print(f"  ✅ Historical volatility {window} created")
    
    # 2. EWMA volatility (exponentially weighted)
    for span in [10, 20, 60]:
        ewma_vol = returns.ewm(span=span, adjust=False).std() * np.sqrt(252)
        feature_df[f'{asset}_ewma_vol_{span}'] = ewma_vol
        print(f"  ✅ EWMA volatility {span} created")
    
    # 3. Realized volatility (Parkinson's estimator)
    # Using close price only as proxy
    feature_df[f'{asset}_realized_vol'] = returns.abs().rolling(20).mean() * np.sqrt(252)
    print(f"  ✅ Realized volatility created")
    
    # 4. Volatility of volatility
    for window in [20, 60]:
        vol = returns.rolling(window).std()
        feature_df[f'{asset}_vol_of_vol_{window}'] = vol.rolling(window).std()
        print(f"  ✅ Vol of vol {window} created")
    
    # 5. Volatility ratio (short-term vs long-term)
    feature_df[f'{asset}_vol_ratio_5_20'] = feature_df[f'{asset}_hist_vol_5'] / feature_df[f'{asset}_hist_vol_20']
    feature_df[f'{asset}_vol_ratio_10_60'] = feature_df[f'{asset}_hist_vol_10'] / feature_df[f'{asset}_hist_vol_60']
    print(f"  ✅ Volatility ratios created")
    
    print(f"✅ Created {len(feature_df.columns)} volatility features")
    return feature_df

# Create volatility features for SPY
spy_vol_features = create_volatility_features(returns_data, 'SPY')
print(f"\nSPY Volatility Features Preview:")
print(spy_vol_features.head())
```

### Volatility Features focus entirely on measuring the speed, intensity, and acceleration of market fear.

Historical Volatility (_hist_vol_)
What it is: The annualized standard deviation of returns across 5 to 60 days.\
Why you need it: This feeds our Volatility Clustering rules directly into the computer's memory matrix. It acts as the standard baseline gauge of market panic.

EWMA Volatility (_ewma_vol_)\
What it is: Exponentially Weighted Moving Average volatility.\
Why it matters: Standard historical volatility treats a market crash from 19 days ago with the exact same importance as a crash that happened this morning.
##### EWMA fixes this flaw. It places massive mathematical weight on the most recent hours of trading, giving our AI an ultra-fast, reactive risk sensor that triggers at the absolute first sign of a sell-off.

Realized Volatility Proxy (_realized_vol)\
What it is: Takes the absolute value of daily returns  and calculates a 20-day moving average.\
Why it matters: Taking the absolute value turns a  crash and a  rally into the exact same value: a raw .
It ignores whether the market went up or down and purely measures the total electrical energy pumping through the stock exchange.

Volatility of Volatility (_vol_of_vol_)\
What it is: The standard deviation of your standard deviation.\
Why it matters: This measures risk acceleration.If volatility rises slowly from 12% to 14%, Vol-of-Vol stays low (calm, predictable growth in risk).If volatility explodes from 12% to 45% in two days, Vol-of-Vol goes off the charts. It warns the model that the market environment is shifting gears into a chaotic state.

Volatility Ratios (_vol_ratio_)\
What it is: Compares short-term panic against long-term historical context (e.g., dividing 5-day risk by 20-day risk).\
Why it matters:If the ratio is greater than 1.0, it proves that short-term panic is spiking out of nowhere compared to the usual baseline. This signals a fresh, incoming market shock.If the ratio drops below 1.0, it proves the recent shock is deflating and the market is cooling down.

```python
# Cell 7: Calendar Features
def create_calendar_features(data):
    """
    Create calendar-based features
    """
    print("\nCreating calendar features...")
    
    feature_df = pd.DataFrame(index=data.index)
    
    # Basic calendar features
    feature_df['day_of_week'] = data.index.dayofweek
    feature_df['month'] = data.index.month
    feature_df['quarter'] = data.index.quarter
    feature_df['year'] = data.index.year
    feature_df['day_of_year'] = data.index.dayofyear
    
    # Is it month end?
    feature_df['is_month_end'] = data.index.is_month_end.astype(int)
    feature_df['is_month_start'] = data.index.is_month_start.astype(int)
    feature_df['is_quarter_end'] = data.index.is_quarter_end.astype(int)
    feature_df['is_year_end'] = data.index.is_year_end.astype(int)
    
    # Days since/until
    feature_df['days_in_month'] = data.index.days_in_month
    feature_df['days_until_month_end'] = data.index.days_in_month - data.index.day
    
    # Week number
    feature_df['week_of_year'] = data.index.isocalendar().week
    
    # Create cyclical features for day of week and month
    feature_df['day_sin'] = np.sin(2 * np.pi * feature_df['day_of_week'] / 7)
    feature_df['day_cos'] = np.cos(2 * np.pi * feature_df['day_of_week'] / 7)
    feature_df['month_sin'] = np.sin(2 * np.pi * feature_df['month'] / 12)
    feature_df['month_cos'] = np.cos(2 * np.pi * feature_df['month'] / 12)
    
    print(f"✅ Created {len(feature_df.columns)} calendar features")
    return feature_df

# Create calendar features
calendar_features = create_calendar_features(returns_data)
print(f"\nCalendar Features Preview:")
print(calendar_features.head())
```

 #### Standard Date Slicing (day_of_week, month, etc.)\
 What: It extracts the numeric day, week, month, and year from your spreadsheet index.\
 Why: Helps your trading engine identify seasonal habits, like the "January Effect" (when stocks historically tend to rise as investors re-enter the market).
 
#### Wall Street Milestones (is_month_end, is_quarter_end, etc.)\
 What: Binary indicators (0 or 1) marking structural calendar deadlines.\
 Why: Large institutional mutual funds and pension funds are legally forced to rebalance their portfolios at the exact end of a month or quarter. This rebalancing causes massive, predictable trading volumes and price shifts.
 
#### The Math Paradox: Cyclical Transformation (_sin and _cos)
 What: Converts day numbers (0 to 6) and month numbers (1 to 12) into continuous wave values using trigonometry.\
 Why (Critical Note): If you feed raw numbers into an AI, it thinks Month 12 (December) is far away from Month 1 (January) because 12 is much bigger than 1. \
 This breaks time-series logic. By wrapping the dates around a circle using Sine and Cosine waves, Month 12 and Month 1 are placed mathematically right next to each other, allowing our machine learning model to understand that time loops seamlessly.

```python
# Cell 8: Cross-Asset Features
def create_cross_asset_features(master_data, returns_data):
    """
    Create features that capture relationships between assets
    """
    print("\nCreating cross-asset features...")
    
    feature_df = pd.DataFrame(index=master_data.index)
    
    # Price ratios (relative strength)
    feature_df['SPY_QQQ_ratio'] = master_data['SPY'] / master_data['QQQ']
    feature_df['SPY_TLT_ratio'] = master_data['SPY'] / master_data['TLT']
    feature_df['SPY_GLD_ratio'] = master_data['SPY'] / master_data['GLD']
    feature_df['SPY_USO_ratio'] = master_data['SPY'] / master_data['USO']
    print(f"  ✅ Price ratios created")
    
    # Return differences (spreads)
    feature_df['SPY_QQQ_spread'] = returns_data['SPY_ret'] - returns_data['QQQ_ret']
    feature_df['SPY_TLT_spread'] = returns_data['SPY_ret'] - returns_data['TLT_ret']
    feature_df['SPY_GLD_spread'] = returns_data['SPY_ret'] - returns_data['GLD_ret']
    feature_df['SPY_USO_spread'] = returns_data['SPY_ret'] - returns_data['USO_ret']
    print(f"  ✅ Return spreads created")
    
    # Rolling correlations
    for window in [20, 60]:
        feature_df[f'corr_SPY_QQQ_{window}'] = returns_data['SPY_ret'].rolling(window).corr(returns_data['QQQ_ret'])
        feature_df[f'corr_SPY_TLT_{window}'] = returns_data['SPY_ret'].rolling(window).corr(returns_data['TLT_ret'])
        feature_df[f'corr_SPY_GLD_{window}'] = returns_data['SPY_ret'].rolling(window).corr(returns_data['GLD_ret'])
        feature_df[f'corr_SPY_USO_{window}'] = returns_data['SPY_ret'].rolling(window).corr(returns_data['USO_ret'])
    print(f"  ✅ Rolling correlations created")
    
    # Volatility ratios
    for asset in ['QQQ', 'TLT', 'GLD', 'USO']:
        vol_spy = returns_data['SPY_ret'].rolling(20).std() * np.sqrt(252)
        vol_asset = returns_data[f'{asset}_ret'].rolling(20).std() * np.sqrt(252)
        feature_df[f'vol_ratio_SPY_{asset}'] = vol_spy / vol_asset
    print(f"  ✅ Volatility ratios created")
    
    print(f"✅ Created {len(feature_df.columns)} cross-asset features")
    return feature_df

# Create cross-asset features
cross_features = create_cross_asset_features(master_data, returns_data)
print(f"\nCross-Asset Features Preview:")
print(cross_features.head())
```

```python
# Cell 9: Regime Features
def create_regime_features(returns_data, asset, vol_series):
    """
    Create regime-based features
    """
    print(f"\nCreating regime features for {asset}...")
    
    feature_df = pd.DataFrame(index=vol_series.index)
    
    # Simple threshold-based regimes
    q33 = vol_series.quantile(0.33)
    q67 = vol_series.quantile(0.67)
    
    # Create regime labels
    feature_df[f'{asset}_regime'] = 0  # Default
    feature_df.loc[vol_series <= q33, f'{asset}_regime'] = 0  # Low vol
    feature_df.loc[(vol_series > q33) & (vol_series <= q67), f'{asset}_regime'] = 1  # Medium vol
    feature_df.loc[vol_series > q67, f'{asset}_regime'] = 2  # High vol
    
    # One-hot encoded regimes
    feature_df[f'{asset}_regime_low'] = (feature_df[f'{asset}_regime'] == 0).astype(int)
    feature_df[f'{asset}_regime_medium'] = (feature_df[f'{asset}_regime'] == 1).astype(int)
    feature_df[f'{asset}_regime_high'] = (feature_df[f'{asset}_regime'] == 2).astype(int)
    
    # Volatility acceleration (change in regime)
    feature_df[f'{asset}_regime_change'] = feature_df[f'{asset}_regime'].diff()
    
    # Days in current regime
    regime_series = feature_df[f'{asset}_regime']
    days_in_regime = np.zeros(len(regime_series))
    current_regime = None
    count = 0
    
    for i, regime in enumerate(regime_series):
        if regime == current_regime:
            count += 1
        else:
            current_regime = regime
            count = 1
        days_in_regime[i] = count
    
    feature_df[f'{asset}_days_in_regime'] = days_in_regime
    
    print(f"✅ Created {len(feature_df.columns)} regime features")
    return feature_df

# First get SPY volatility
spy_vol_20 = returns_data['SPY_ret'].rolling(20).std() * np.sqrt(252)
spy_regime_features = create_regime_features(returns_data, 'SPY', spy_vol_20)
print(f"\nSPY Regime Features Preview:")
print(spy_regime_features.head())
```

```python
# Cell 10: Combine All Features
def combine_all_features(asset, returns_data, master_data, vol_series):
    """
    Combine all features into a single DataFrame
    """
    print(f"\n{'='*60}")
    print(f"COMBINING ALL FEATURES FOR {asset}")
    print('='*60)
    
    # Create all feature sets
    print("\n1. Creating lag features...")
    lag_features = create_lag_features(returns_data, asset)
    
    print("\n2. Creating rolling features...")
    rolling_features = create_rolling_features(returns_data, asset)
    
    print("\n3. Creating technical indicators...")
    tech_features = create_technical_indicators(returns_data, asset)
    
    print("\n4. Creating volatility features...")
    vol_features = create_volatility_features(returns_data, asset)
    
    print("\n5. Creating regime features...")
    regime_features = create_regime_features(returns_data, asset, vol_series)
    
    print("\n6. Adding calendar features...")
    calendar_features = create_calendar_features(returns_data)
    
    print("\n7. Adding cross-asset features...")
    cross_features = create_cross_asset_features(master_data, returns_data)
    
    # Combine all features
    print("\n8. Combining all features...")
    all_features = pd.DataFrame(index=returns_data.index)
    
    # Add each feature set
    feature_sets = [
        lag_features,
        rolling_features,
        tech_features,
        vol_features,
        regime_features,
        calendar_features,
        cross_features
    ]
    
    for features in feature_sets:
        if features is not None and not features.empty:
            # Only add columns that aren't already in all_features
            new_cols = [col for col in features.columns if col not in all_features.columns]
            if new_cols:
                all_features[new_cols] = features[new_cols]
    
    print(f"✅ Combined feature set: {all_features.shape}")
    print(f"   Rows: {all_features.shape[0]}, Columns: {all_features.shape[1]}")
    
    # Preview
    print("\n📊 Feature Preview:")
    print(all_features.head())
    
    print("\n📊 Feature Statistics:")
    print(all_features.describe().round(4))
    
    return all_features

# Get SPY volatility series
spy_vol_20 = returns_data['SPY_ret'].rolling(20).std() * np.sqrt(252)

# Combine all features
spy_features = combine_all_features(
    'SPY', 
    returns_data, 
    master_data, 
    spy_vol_20
)
```

```python
# Cell 11: Create Target Variable
def create_target_variable(returns_data, asset, horizon=1):
    """
    Create target variable: future volatility
    """
    print(f"\nCreating target variable for {asset}...")
    
    ret_col = f'{asset}_ret'
    if ret_col not in returns_data.columns:
        print(f"❌ {ret_col} not found!")
        return None
    
    # Future realized volatility
    # Using 5-day future volatility as target
    future_vol = returns_data[ret_col].rolling(5).std().shift(-5) * np.sqrt(252)
    target = future_vol
    
    print(f"✅ Target variable created: {target.name}")
    print(f"   Shape: {target.shape}")
    print(f"   Mean: {target.mean():.4f}")
    print(f"   Std: {target.std():.4f}")
    print(f"   Min: {target.min():.4f}")
    print(f"   Max: {target.max():.4f}")
    
    return target

# Create target
target = create_target_variable(returns_data, 'SPY')
```

```python
# Cell 12: Clean and Prepare Final Dataset
def prepare_final_dataset(features, target):
    """
    Prepare final dataset for modeling
    """
    print("\n" + "="*60)
    print("PREPARING FINAL DATASET")
    print("="*60)
    
    # Combine features and target
    dataset = features.copy()
    dataset['target'] = target
    
    # Remove rows with NaN
    print(f"\nBefore cleaning: {dataset.shape}")
    dataset_clean = dataset.dropna()
    print(f"After cleaning: {dataset_clean.shape}")
    print(f"Removed rows: {dataset.shape[0] - dataset_clean.shape[0]}")
    
    # Split into features and target
    X = dataset_clean.drop('target', axis=1)
    y = dataset_clean['target']
    
    print(f"\nFeature set: {X.shape}")
    print(f"Target set: {y.shape}")
    
    return X, y, dataset_clean

# Prepare final dataset
X, y, dataset_clean = prepare_final_dataset(spy_features, target)
```

```python
# Cell 13: Feature Analysis
def analyze_features(X):
    """
    Analyze feature characteristics
    """
    print("\n" + "="*60)
    print("FEATURE ANALYSIS")
    print("="*60)
    
    print(f"\n1. Feature Count: {len(X.columns)}")
    
    # Feature types
    feature_types = {
        'Lag': [col for col in X.columns if 'lag' in col.lower()],
        'Rolling': [col for col in X.columns if 'roll' in col.lower()],
        'Technical': [col for col in X.columns if any(x in col for x in ['RSI', 'MACD', 'BB', 'MA', 'EMA', 'ATR'])],
        'Volatility': [col for col in X.columns if 'vol' in col.lower()],
        'Calendar': [col for col in X.columns if any(x in col for x in ['day', 'month', 'quarter', 'year', 'week'])],
        'Cross': [col for col in X.columns if any(x in col for x in ['ratio', 'spread', 'corr'])],
        'Regime': [col for col in X.columns if 'regime' in col.lower()]
    }
    
    print("\n2. Feature Distribution by Type:")
    for ftype, cols in feature_types.items():
        if cols:
            print(f"   {ftype}: {len(cols)} features")
    
    # Feature statistics
    print("\n3. Feature Statistics:")
    print(f"   Mean correlation with target: {X.corrwith(y).mean():.4f}")
    print(f"   Max correlation with target: {X.corrwith(y).max():.4f}")
    print(f"   Min correlation with target: {X.corrwith(y).min():.4f}")
    
    # Top features by correlation
    print("\n4. Top 10 Features by Correlation with Target:")
    correlations = X.corrwith(y).abs().sort_values(ascending=False)
    for i, (feature, corr) in enumerate(correlations.head(10).items(), 1):
        print(f"   {i:2}. {feature:30} {corr:.4f}")
    
    # Feature correlation matrix
    print("\n5. Feature Correlation Matrix:")
    corr_matrix = X.corr()
    high_corr_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > 0.9:
                high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))
    
    if high_corr_pairs:
        print(f"   Found {len(high_corr_pairs)} highly correlated pairs (|corr| > 0.9):")
        for pair in high_corr_pairs[:10]:
            print(f"     {pair[0]} ↔ {pair[1]}: {pair[2]:.4f}")
    else:
        print("   No highly correlated pairs found")
    
    return feature_types, correlations

# Analyze features
feature_types, top_correlations = analyze_features(X)
```

```python
# Cell 14: Save Feature Dataset
def save_features(X, y, dataset_clean, timestamp=None):
    """
    Save feature datasets
    """
    if timestamp is None:
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    
    # Save features
    X.to_csv(f'features_X_{timestamp}.csv')
    print(f"✅ Features saved: features_X_{timestamp}.csv")
    
    # Save target
    y.to_csv(f'target_y_{timestamp}.csv')
    print(f"✅ Target saved: target_y_{timestamp}.csv")
    
    # Save full dataset
    dataset_clean.to_csv(f'full_dataset_{timestamp}.csv')
    print(f"✅ Full dataset saved: full_dataset_{timestamp}.csv")
    
    # Save feature list
    feature_list = pd.DataFrame({
        'Feature': X.columns,
        'Type': 'Unknown'
    })
    feature_list.to_csv(f'feature_list_{timestamp}.csv', index=False)
    print(f"✅ Feature list saved: feature_list_{timestamp}.csv")
    
    return timestamp

# Save feature dataset
timestamp = save_features(X, y, dataset_clean)
```
