```python
# Cell 7: Data Cleaning Function
def clean_price_data(df, symbol):
    """
    Clean and prepare price data
    Handles: Missing values, outliers, splits, dividends
    """
    df_clean = df.copy()
    
    print(f"\nCleaning {symbol}...")
    print("-"*30)
    
    # 1. Check for missing values
    missing = df_clean.isnull().sum()
    if missing.sum() > 0:
        print(f"  ⚠️ Found {missing.sum()} missing values")
        # Forward fill then backward fill for small gaps
        df_clean = df_clean.fillna(method='ffill').fillna(method='bfill')
        print(f"  ✅ Filled missing values")
    else:
        print(f"  ✅ No missing values")
    
    # 2. Check for outliers using Z-score method
    for col in ['Close', 'High', 'Low']:
        if col in df_clean.columns:
            z_scores = np.abs(stats.zscore(df_clean[col].fillna(method='ffill')))
            outliers = (z_scores > 4).sum()
            if outliers > 0:
                print(f"  ⚠️ Found {outliers} outliers in {col}")
    
    # 3. Calculate returns
    df_clean['Returns'] = df_clean['Close'].pct_change()
    
    # 4. Calculate log returns (better for volatility)
    df_clean['Log_Returns'] = np.log(df_clean['Close'] / df_clean['Close'].shift(1))
    
    # 5. Add basic rolling statistics
    df_clean['MA5'] = df_clean['Close'].rolling(5).mean()
    df_clean['MA20'] = df_clean['Close'].rolling(20).mean()
    df_clean['MA50'] = df_clean['Close'].rolling(50).mean()
    df_clean['MA200'] = df_clean['Close'].rolling(200).mean()
    df_clean['Rolling_Vol_20'] = df_clean['Returns'].rolling(20).std() * np.sqrt(252)
    df_clean['Rolling_Vol_60'] = df_clean['Returns'].rolling(60).std() * np.sqrt(252)
    
    print(f"  ✅ Data cleaning complete")
    print(f"  📊 {len(df_clean)} rows, {len(df_clean.columns)} columns")
    
    return df_clean

# Clean SPY data
spy_clean = clean_price_data(spy, 'SPY')
print("\n" + "="*50)
print(spy_clean.head())
```

## test for stationarity 

```python
# Cell 8: Stationarity Tests
from statsmodels.tsa.stattools import adfuller, kpss

def check_stationarity(series, series_name, alpha=0.05):
    """
    Perform ADF and KPSS tests for stationarity
    """
    print(f"\nStationarity Tests for {series_name}")
    print("="*50)
    
    # Remove NaN values
    series_clean = series.dropna()
    
    if len(series_clean) < 10:
        print("   Not enough data points for stationarity test")
        return False
    
    # ADF Test
    adf_result = adfuller(series_clean, autolag='AIC')
    print(f"ADF Test (H0: Series has unit root - non-stationary):")
    print(f"  Test Statistic: {adf_result[0]:.4f}")
    print(f"  p-value: {adf_result[1]:.4f}")
    print(f"  Critical Values:")
    for key, value in adf_result[4].items():
        print(f"    {key}: {value:.4f}")
    
    adf_stationary = adf_result[1] < alpha
    if adf_stationary:
        print(f"   Reject H0: Series is stationary (p < {alpha})")
    else:
        print(f"   Fail to reject H0: Series is non-stationary (p >= {alpha})")
    
    # KPSS Test
    kpss_result = kpss(series_clean, regression='c', nlags='auto')
    print(f"\nKPSS Test (H0: Series is stationary):")
    print(f"  Test Statistic: {kpss_result[0]:.4f}")
    print(f"  p-value: {kpss_result[1]:.4f}")
    print(f"  Critical Values:")
    for key, value in kpss_result[3].items():
        print(f"    {key}: {value:.4f}")
    
    kpss_stationary = kpss_result[1] >= alpha
    if kpss_stationary:
        print(f"   Fail to reject H0: Series is stationary (p >= {alpha})")
    else:
        print(f"   Reject H0: Series is non-stationary (p < {alpha})")
    
    # Conclusion
    if adf_stationary and kpss_stationary:
        print("\n Conclusion: Series is STRONGLY STATIONARY")
        return True
    elif adf_stationary and not kpss_stationary:
        print("\n Conclusion: Mixed results - Series may be trend stationary")
        return True
    elif not adf_stationary and kpss_stationary:
        print("\n Conclusion: Mixed results - Series may have unit root")
        return False
    else:
        print("\n Conclusion: Series is STRONGLY NON-STATIONARY")
        return False

# Test stationarity of SPY prices and returns
spy_stationary = check_stationarity(spy_clean['Close'], 'SPY Close Price')
returns_stationary = check_stationarity(spy_clean['Log_Returns'].dropna(), 'SPY Log Returns')
```

### KPSS test :- complete opposite of ADF ( default assumption :->i.e null :- data is stationary)

here check for both test stationarity condition and then give answer

```python
# Cell 9: Differencing to Make Series Stationary
def difference_series(series, order=1):
    """
    Apply differencing to make series stationary
    """
    diff = series.diff(order)
    return diff

# Apply differencing
spy_diff_1 = difference_series(spy_clean['Close'], 1)
spy_diff_1 = spy_diff_1.dropna()

# Check stationarity of differenced series
print("After 1st Differencing:")
check_stationarity(spy_diff_1, 'SPY Close Price - 1st Difference')

# Visualize before and after
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(spy_clean.index, spy_clean['Close'], color='blue', linewidth=1)
axes[0].set_title('Original SPY Prices')
axes[0].set_ylabel('Price ($)')
axes[0].grid(True, alpha=0.3)

axes[1].plot(spy_diff_1.index, spy_diff_1, color='red', linewidth=1)
axes[1].axhline(y=0, color='black', linestyle='--', linewidth=0.5)
axes[1].set_title('SPY Prices - 1st Difference')
axes[1].set_ylabel('Price Change ($)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 10: Box-Cox Transformation
from scipy import stats

def apply_boxcox(series):
    """
    Apply Box-Cox transformation to make data more normal
    """
    # Remove zeros and negatives (Box-Cox requires positive values)
    series_positive = series[series > 0].dropna()
    
    if len(series_positive) > 0:
        try:
            # Apply Box-Cox
            transformed, lambda_param = stats.boxcox(series_positive)
            print(f"  Box-Cox lambda: {lambda_param:.4f}")
            return transformed, lambda_param
        except Exception as e:
            print(f"  ⚠️ Box-Cox failed: {e}")
            return series, None
    else:
        print("  ⚠️ Cannot apply Box-Cox (non-positive values)")
        return series, None

# Apply to SPY prices
prices = spy_clean['Close']
transformed_prices, lambda_param = apply_boxcox(prices)

# Visualize before and after
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Before
axes[0].hist(prices.dropna(), bins=50, color='blue', alpha=0.7, edgecolor='black')
axes[0].set_title('Original SPY Prices')
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Frequency')
axes[0].axvline(x=prices.mean().squeeze(), color='red', linestyle='--', label=f'Mean: ${prices.mean().squeeze():.2f}')
axes[0].legend()

# After
if transformed_prices is not None and not isinstance(transformed_prices, pd.Series):
    axes[1].hist(transformed_prices, bins=50, color='green', alpha=0.7, edgecolor='black')
    lbl = f'λ={lambda_param:.2f}' if lambda_param is not None else 'Estimated'
    axes[1].set_title(f'Box-Cox Transformed ({lbl})')
    axes[1].set_xlabel('Transformed Price')
    axes[1].set_ylabel('Frequency')
else:
    axes[1].hist(prices.dropna(), bins=50, color='green', alpha=0.7, edgecolor='black')
    axes[1].set_title('No Transformation Applied')

plt.tight_layout()
plt.show()
```

```python

def create_master_dataset(raw_data):
  
    print("Creating master dataset...")
    print("-"*40)
    
    # Start with SPY
    if 'SPY' in raw_data and raw_data['SPY'] is not None:
        master = raw_data['SPY'][['Close']].copy()
        master.columns = ['SPY']
    else:
        print("❌ SPY data missing!")
        return None
    
    # Add other assets
    for name in ['QQQ', 'TLT', 'GLD', 'USO', 'DXY']:
        if name in raw_data and raw_data[name] is not None:
            # Align data to SPY dates
            temp = raw_data[name]['Close'].copy()
            # Reindex to SPY dates
            temp = temp.reindex(master.index)
            master[name] = temp
            print(f"   Added {name}")
        else:
            print(f"   {name} data missing, skipping")
    
    # Add VIX separately (it's a volatility index)
    if 'VIX' in raw_data and raw_data['VIX'] is not None:
        vix = raw_data['VIX']['Close'].reindex(master.index)
        master['VIX'] = vix
        print(f"  Added VIX")
    else:
        print(f"   VIX data missing, skipping")
    
    print("-"*40)
    print(f" Master dataset created with {len(master.columns)} assets")
    print(f" {len(master)} rows, {master.index[0]} to {master.index[-1]}")
    print(f"Assets: {', '.join(master.columns)}")
    
    return master

# Create master dataset
master_data = create_master_dataset(raw_data)
print("\nMaster Data Preview:")
print(master_data.head())
```

### We designed a centralized consolidation function to merge our 7 distinct asset data streams into a singular unified database (master_data). Crucially, we anchored the timeline index to our primary asset (SPY), and utilized Pandas reindexing to mathematically align the trading calendars of all satellite assets (QQQ, TLT, GLD, USO, DXY, VIX). This calendar synchronization ensures that every row represents the exact same trading day across the global economy, preventing any future data leakage during forecasting and testing.


No, the date is not deleted!\
When you use reindex, Python preserves the date and handles missing data by placing a blank placeholder (NaN, which means Not a Number) in that specific cell.\
What Reindexing Actually Does: Think of reindex like setting up an attendance sheet where the days of the week are written by SPY.

```python
# Cell 12: Handle Missing Values in Master Dataset
def handle_missing_data(df):
    """
    Handle missing values in the master dataset
    """
    print("\nHandling Missing Values...")
    print("-"*40)
    
    # Check missing values
    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    
    print("Missing values per column:")
    for col in df.columns:
        if missing[col] > 0:
            print(f"  {col}: {missing[col]:5} ({missing_pct[col]:.1f}%)")
    
    # Forward fill for small gaps
    df_filled = df.fillna(method='ffill')
    df_filled = df_filled.fillna(method='bfill')
    
    # If still missing, use median
    for col in df_filled.columns:
        if df_filled[col].isnull().sum() > 0:
            df_filled[col] = df_filled[col].fillna(df_filled[col].median())
    
    new_missing = df_filled.isnull().sum().sum()
    print(f"\n Missing values after handling: {new_missing}")
    
    return df_filled

# Handle missing values
master_data_clean = handle_missing_data(master_data)
print("\n" + "="*50)
print(master_data_clean.head())
```

```python
# Cell 14: Calculate Rolling Volatility for All Assets
def calculate_rolling_volatility(data, window=20):
    """
    Calculate rolling volatility for all assets
    """
    vol_df = pd.DataFrame(index=data.index)
    
    for col in data.columns:
        # Calculate returns
        returns = data[col].pct_change()
        
        # Rolling volatility (annualized)
        vol = returns.rolling(window).std() * np.sqrt(252)
        vol_df[col] = vol
    
    return vol_df

# Calculate different windows
vol_20 = calculate_rolling_volatility(master_data_clean, 20)
vol_60 = calculate_rolling_volatility(master_data_clean, 60)

# Visualize SPY volatility
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Price vs Volatility
ax1 = axes[0]
ax1.plot(master_data_clean.index, master_data_clean['SPY'], color='blue', linewidth=1, label='SPY Price')
ax1.set_ylabel('Price ($)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)

ax1_2 = ax1.twinx()
ax1_2.plot(vol_20.index, vol_20['SPY'] * 100, color='red', linewidth=1, label='20-day Vol')
ax1_2.plot(vol_60.index, vol_60['SPY'] * 100, color='orange', linewidth=1, label='60-day Vol')
ax1_2.set_ylabel('Volatility (%)', color='red')
ax1_2.tick_params(axis='y', labelcolor='red')
ax1_2.legend(loc='upper right')
ax1_2.set_title('SPY Price and Rolling Volatility')

# All assets volatility
ax2 = axes[1]
for col in vol_20.columns:
    ax2.plot(vol_20.index, vol_20[col] * 100, linewidth=1, label=col, alpha=0.7)
ax2.set_ylabel('20-day Volatility (%)')
ax2.set_title('Rolling Volatility Comparison (20-day)')
ax2.legend(loc='upper right')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 17: Quick Validation Function
def validate_data_integrity(data):
    """
    Validate data integrity with simple checks
    """
    print("\n🔍 Data Integrity Check")
    print("="*50)
    
    checks_passed = 0
    total_checks = 6
    
    # Check 1: No negative prices (except VIX which can't be negative)
    for col in data.columns:
        if col != 'VIX' and col != 'DXY':  # DXY can be negative?
            negative = (data[col] < 0).sum()
            if negative == 0:
                print(f"✅ {col}: No negative prices")
            else:
                print(f"❌ {col}: {negative} negative values found")
    checks_passed += 1
    
    # Check 2: Data types
    all_numeric = all(pd.api.types.is_numeric_dtype(data[col]) for col in data.columns)
    print(f"✅ All columns are numeric: {all_numeric}")
    checks_passed += 1 if all_numeric else 0
    
    # Check 3: Date is sorted
    is_sorted = data.index.is_monotonic_increasing
    print(f"✅ Index is sorted: {is_sorted}")
    checks_passed += 1 if is_sorted else 0
    
    # Check 4: No extreme outliers (beyond 5 standard deviations)
    outlier_count = 0
    for col in data.columns:
        mean = data[col].mean()
        std = data[col].std()
        outliers = ((data[col] - mean).abs() > 5 * std).sum()
        outlier_count += outliers
        pct = (outliers / len(data)) * 100
        if pct < 1:
            print(f"✅ {col}: {outliers} outliers ({pct:.2f}%)")
        else:
            print(f"⚠️ {col}: {outliers} outliers ({pct:.2f}%)")
    checks_passed += 1 if outlier_count == 0 else 0
    
    # Check 5: Complete data for recent periods
    recent_data = data.tail(252)  # Last year
    complete = (recent_data.count() / len(recent_data) * 100)
    if (complete > 95).all():
        checks_passed += 1
        print(f"✅ Recent data completeness: {complete.min():.1f}%")
    
    # Check 6: No NaN values
    no_nan = data.isnull().sum().sum() == 0
    print(f"✅ No NaN values: {no_nan}")
    checks_passed += 1 if no_nan else 0
    
    print(f"\n📊 Overall: {checks_passed}/{total_checks} checks passed")
    
    return checks_passed == total_checks

# Run validation
is_valid = validate_data_integrity(master_data_clean)
```
