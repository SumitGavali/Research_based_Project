```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')


import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

```


XGBoost stands for Extreme Gradient Boosting. It is a highly efficient, scalable machine learning library designed for speed and performance.






LightGBM stands for Light Gradient Boosting Machine. It is a high-performance, open-source gradient boosting framework developed by Microsoft.






```python
# Cell 2: Load Feature Data
def load_feature_data():
    
    import glob
    
    # Find latest files
    feature_files = glob.glob('features_X_*.csv')
    target_files = glob.glob('target_y_*.csv')
    
    if feature_files and target_files:
        latest_features = sorted(feature_files)[-1]
        latest_target = sorted(target_files)[-1]
        
        X = pd.read_csv(latest_features, index_col=0, parse_dates=True)
        y = pd.read_csv(latest_target, index_col=0, parse_dates=True)
        
        print(f"Loaded features: {latest_features}")
        print(f"Loaded target: {latest_target}")
        print(f"Features shape: {X.shape}")
        print(f"Target shape: {y.shape}")
        
        return X, y
    else:
        print("❌ Feature files not found!")
        return None, None

# Load data
X, y = load_feature_data()
```

```python
# Cell 3: Load GARCH Results for Comparison
def load_garch_results():
    """Load GARCH forecast from Phase 4"""
    import glob
    
    garch_files = glob.glob('garch_forecast_*.csv')
    if garch_files:
        latest_garch = sorted(garch_files)[-1]
        garch_forecast = pd.read_csv(latest_garch, index_col=0, parse_dates=True)
        print(f" Loaded GARCH forecast: {latest_garch}")
        return garch_forecast
    else:
        print(" No GARCH forecast found")
        return None

# Load GARCH results
garch_forecast = load_garch_results()
```

```python
# Cell 4: Prepare Data for ML
def prepare_ml_data(X, y, test_size=0.2):
    """
    Prepare data for machine learning with time series split
    """
    print("\n" + "="*60)
    print("PREPARING DATA FOR MACHINE LEARNING")
    print("="*60)
    
    # Align X and y
    common_idx = X.index.intersection(y.index)
    X_aligned = X.loc[common_idx]
    y_aligned = y.loc[common_idx]
    
    print(f"Aligned data: {len(common_idx)} observations")
    
    # Split by date (not random - time series!)
    split_idx = int(len(X_aligned) * (1 - test_size))
    
    X_train = X_aligned.iloc[:split_idx]
    X_test = X_aligned.iloc[split_idx:]
    y_train = y_aligned.iloc[:split_idx]
    y_test = y_aligned.iloc[split_idx:]
    
    print(f"\nTraining set: {len(X_train)} observations")
    print(f"  Date range: {X_train.index[0]} to {X_train.index[-1]}")
    print(f"Test set: {len(X_test)} observations")
    print(f"  Date range: {X_test.index[0]} to {X_test.index[-1]}")
    
    # Scale features (important for some models)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Convert back to DataFrame for feature importance
    X_train_scaled_df = pd.DataFrame(X_train_scaled, index=X_train.index, columns=X_train.columns)
    X_test_scaled_df = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
    
    return X_train_scaled_df, X_test_scaled_df, y_train, y_test, scaler

# Prepare data
X_train, X_test, y_train, y_test, scaler = prepare_ml_data(X, y)
```

```python
# Cell 5: Train XGBoost Model
def train_xgboost(X_train, y_train, X_test, y_test):
    """
    Train XGBoost model with hyperparameter tuning
    """
    print("\n" + "="*60)
    print("TRAINING XGBOOST MODEL")
    print("="*60)
    
    # Base model
    xgb_model = xgb.XGBRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    
    # Fit model
    xgb_model.fit(X_train, y_train)
    
    # Predictions
    y_pred_train = xgb_model.predict(X_train)
    y_pred_test = xgb_model.predict(X_test)
    
    # Metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
    train_mae = mean_absolute_error(y_train, y_pred_train)
    test_mae = mean_absolute_error(y_test, y_pred_test)
    train_r2 = r2_score(y_train, y_pred_train)
    test_r2 = r2_score(y_test, y_pred_test)
    
    print(f"\nTraining Metrics:")
    print(f"  RMSE: {train_rmse:.4f}")
    print(f"  MAE: {train_mae:.4f}")
    print(f"  R²: {train_r2:.4f}")
    
    print(f"\nTest Metrics:")
    print(f"  RMSE: {test_rmse:.4f}")
    print(f"  MAE: {test_mae:.4f}")
    print(f"  R²: {test_r2:.4f}")
    
    return xgb_model, y_pred_train, y_pred_test, {'train_rmse': train_rmse, 'test_rmse': test_rmse, 
                                                   'train_mae': train_mae, 'test_mae': test_mae,
                                                   'train_r2': train_r2, 'test_r2': test_r2}

# Train XGBoost
xgb_model, xgb_pred_train, xgb_pred_test, xgb_metrics = train_xgboost(X_train, y_train, X_test, y_test)
```

```python
# Cell 6: Train LightGBM Model
def train_lightgbm(X_train, y_train, X_test, y_test):
    """
    Train LightGBM model with hyperparameter tuning
    """
    print("\n" + "="*60)
    print("TRAINING LIGHTGBM MODEL")
    print("="*60)
    
    # Base model
    lgb_model = lgb.LGBMRegressor(
        n_estimators=200,
        learning_rate=0.05,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    
    # Fit model
    lgb_model.fit(X_train, y_train)
    
    # Predictions
    y_pred_train = lgb_model.predict(X_train)
    y_pred_test = lgb_model.predict(X_test)
    
    # Metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
    train_mae = mean_absolute_error(y_train, y_pred_train)
    test_mae = mean_absolute_error(y_test, y_pred_test)
    train_r2 = r2_score(y_train, y_pred_train)
    test_r2 = r2_score(y_test, y_pred_test)
    
    print(f"\nTraining Metrics:")
    print(f"  RMSE: {train_rmse:.4f}")
    print(f"  MAE: {train_mae:.4f}")
    print(f"  R²: {train_r2:.4f}")
    
    print(f"\nTest Metrics:")
    print(f"  RMSE: {test_rmse:.4f}")
    print(f"  MAE: {test_mae:.4f}")
    print(f"  R²: {test_r2:.4f}")
    
    return lgb_model, y_pred_train, y_pred_test, {'train_rmse': train_rmse, 'test_rmse': test_rmse,
                                                   'train_mae': train_mae, 'test_mae': test_mae,
                                                   'train_r2': train_r2, 'test_r2': test_r2}

# Train LightGBM
lgb_model, lgb_pred_train, lgb_pred_test, lgb_metrics = train_lightgbm(X_train, y_train, X_test, y_test)
```

```python
# Cell 7: Train Random Forest Model
def train_randomforest(X_train, y_train, X_test, y_test):
    """
    Train Random Forest model
    """
    print("\n" + "="*60)
    print("TRAINING RANDOM FOREST MODEL")
    print("="*60)
    
    # Base model
    rf_model = RandomForestRegressor(
        n_estimators=200,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42
    )
    
    # Fit model
    rf_model.fit(X_train, y_train)
    
    # Predictions
    y_pred_train = rf_model.predict(X_train)
    y_pred_test = rf_model.predict(X_test)
    
    # Metrics
    train_rmse = np.sqrt(mean_squared_error(y_train, y_pred_train))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
    train_mae = mean_absolute_error(y_train, y_pred_train)
    test_mae = mean_absolute_error(y_test, y_pred_test)
    train_r2 = r2_score(y_train, y_pred_train)
    test_r2 = r2_score(y_test, y_pred_test)
    
    print(f"\nTraining Metrics:")
    print(f"  RMSE: {train_rmse:.4f}")
    print(f"  MAE: {train_mae:.4f}")
    print(f"  R²: {train_r2:.4f}")
    
    print(f"\nTest Metrics:")
    print(f"  RMSE: {test_rmse:.4f}")
    print(f"  MAE: {test_mae:.4f}")
    print(f"  R²: {test_r2:.4f}")
    
    return rf_model, y_pred_train, y_pred_test, {'train_rmse': train_rmse, 'test_rmse': test_rmse,
                                                  'train_mae': train_mae, 'test_mae': test_mae,
                                                  'train_r2': train_r2, 'test_r2': test_r2}

# Train Random Forest
rf_model, rf_pred_train, rf_pred_test, rf_metrics = train_randomforest(X_train, y_train, X_test, y_test)
```

```python
# Cell 8: Compare All Models
def compare_models(y_test, xgb_pred, lgb_pred, rf_pred, garch_forecast=None):
    """
    Compare all models including GARCH
    """
    print("\n" + "="*60)
    print("MODEL COMPARISON")
    print("="*60)
    
    results = []
    
    # XGBoost
    results.append({
        'Model': 'XGBoost',
        'RMSE': np.sqrt(mean_squared_error(y_test, xgb_pred)),
        'MAE': mean_absolute_error(y_test, xgb_pred),
        'R2': r2_score(y_test, xgb_pred)
    })
    
    # LightGBM
    results.append({
        'Model': 'LightGBM',
        'RMSE': np.sqrt(mean_squared_error(y_test, lgb_pred)),
        'MAE': mean_absolute_error(y_test, lgb_pred),
        'R2': r2_score(y_test, lgb_pred)
    })
    
    # Random Forest
    results.append({
        'Model': 'Random Forest',
        'RMSE': np.sqrt(mean_squared_error(y_test, rf_pred)),
        'MAE': mean_absolute_error(y_test, rf_pred),
        'R2': r2_score(y_test, rf_pred)
    })
    
    # GARCH (if available)
    if garch_forecast is not None:
        # Align GARCH forecast with test dates
        garch_aligned = garch_forecast.reindex(y_test.index).dropna()
        y_test_aligned = y_test.loc[garch_aligned.index]
        
        if len(garch_aligned) > 0:
            results.append({
                'Model': 'GARCH(1,1)',
                'RMSE': np.sqrt(mean_squared_error(y_test_aligned, garch_aligned.values.flatten())),
                'MAE': mean_absolute_error(y_test_aligned, garch_aligned.values.flatten()),
                'R2': r2_score(y_test_aligned, garch_aligned.values.flatten())
            })
    
    # Create comparison DataFrame
    comparison_df = pd.DataFrame(results)
    comparison_df = comparison_df.sort_values('RMSE')
    
    print("\nModel Performance Comparison:")
    print("-"*60)
    print(comparison_df.to_string(index=False))
    
    # Best model
    best_model = comparison_df.iloc[0]
    print(f"\n🏆 Best Model: {best_model['Model']}")
    print(f"   RMSE: {best_model['RMSE']:.4f}")
    print(f"   MAE: {best_model['MAE']:.4f}")
    print(f"   R²: {best_model['R2']:.4f}")
    
    return comparison_df

# Compare ML models
comparison_df = compare_models(y_test, xgb_pred_test, lgb_pred_test, rf_pred_test, garch_forecast)
```

More importantly, both have a negative R² score (~ -0.07). A negative $R^2$ means the models performed worse than if you had simply guessed the horizontal line of the historical average for every prediction. They failed to find a pattern.

```python
# Cell 9: Feature Importance Analysis
def analyze_feature_importance(model, feature_names, model_name):
    """
    Analyze and visualize feature importance
    """
    print(f"\n{'='*60}")
    print(f"FEATURE IMPORTANCE: {model_name}")
    print('='*60)
    
    # Get feature importance
    if hasattr(model, 'feature_importances_'):
        importance = model.feature_importances_
        importance = importance / np.sum(importance)
    else:
        print("Model doesn't have feature_importances_ attribute")
        return None
    
    # Create DataFrame
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': importance
    }).sort_values('importance', ascending=False)
    
    # Top 20 features
    print("\nTop 20 Features:")
    print("-"*50)
    for i, row in importance_df.head(20).iterrows():
        print(f"  {row['feature']}: {row['importance']:.4f}")
    
    # Visualization
    fig, axes = plt.subplots(1, 2, figsize=(15, 8))
    
    # Top 20 bar chart
    top_features = importance_df.head(20)
    axes[0].barh(top_features['feature'], top_features['importance'], color='steelblue', alpha=0.7)
    axes[0].set_title(f'Top 20 Features ({model_name})')
    axes[0].set_xlabel('Importance')
    axes[0].invert_yaxis()
    axes[0].grid(True, alpha=0.3)
    
    # Cumulative importance
    cumulative = importance_df['importance'].cumsum()
    axes[1].plot(range(len(cumulative)), cumulative, color='red', linewidth=2)
    axes[1].axhline(y=0.8, color='black', linestyle='--', label='80% threshold')
    axes[1].set_title('Cumulative Feature Importance')
    axes[1].set_xlabel('Number of Features')
    axes[1].set_ylabel('Cumulative Importance')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    # Find number of features to reach 80%
    n_80 = np.where(cumulative >= 0.8)[0][0] + 1
    axes[1].axvline(x=n_80, color='red', linestyle='--', alpha=0.5)
    axes[1].text(n_80, 0.5, f'{n_80} features', rotation=90)
    
    plt.tight_layout()
    plt.show()
    
    return importance_df

# Analyze feature importance for XGBoost
xgb_importance = analyze_feature_importance(xgb_model, X_train.columns, 'XGBoost')

# Analyze feature importance for LightGBM
lgb_importance = analyze_feature_importance(lgb_model, X_train.columns, 'LightGBM')

# Analyze feature importance for Random Forest
rf_importance = analyze_feature_importance(rf_model, X_train.columns, 'Random Forest')
```

```python
# Cell 10: Visualize Predictions (FIXED)
def visualize_predictions(y_test, predictions, model_names, title="Model Predictions"):
    """
    Visualize model predictions vs actual
    """
    print("\n" + "="*60)
    print("PREDICTION VISUALIZATION")
    print("="*60)
    
    # Convert y_test to numpy array if it's a DataFrame
    if hasattr(y_test, 'values'):
        y_test_array = y_test.values.flatten()
        y_test_index = y_test.index
    else:
        y_test_array = y_test.flatten()
        y_test_index = range(len(y_test))
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # Plot 1: Actual vs Predicted (Scatter)
    ax1 = axes[0, 0]
    for name, pred in predictions.items():
        # Ensure pred is flat
        pred_flat = pred.flatten() if hasattr(pred, 'flatten') else pred
        ax1.scatter(y_test_array, pred_flat, alpha=0.3, s=5, label=name)
    ax1.plot([y_test_array.min(), y_test_array.max()], [y_test_array.min(), y_test_array.max()], 
             'k--', linewidth=2, label='Perfect')
    ax1.set_title('Actual vs Predicted Volatility')
    ax1.set_xlabel('Actual Volatility')
    ax1.set_ylabel('Predicted Volatility')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Time Series (Last 252 days)
    ax2 = axes[0, 1]
    last_252 = min(len(y_test_array), 252)
    y_test_last = y_test_array[-last_252:]
    y_test_index_last = y_test_index[-last_252:]
    
    ax2.plot(y_test_index_last, y_test_last, 'k-', linewidth=2, label='Actual')
    for name, pred in predictions.items():
        pred_flat = pred.flatten() if hasattr(pred, 'flatten') else pred
        pred_last = pred_flat[-last_252:]
        ax2.plot(y_test_index_last, pred_last, linewidth=1, alpha=0.7, label=name)
    ax2.set_title('Volatility Forecast (Last Year)')
    ax2.set_ylabel('Volatility (%)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Error Distribution
    ax3 = axes[1, 0]
    for name, pred in predictions.items():
        pred_flat = pred.flatten() if hasattr(pred, 'flatten') else pred
        error = y_test_array - pred_flat
        ax3.hist(error, bins=30, alpha=0.3, label=name, density=True)
    ax3.axvline(x=0, color='black', linestyle='--')
    ax3.set_title('Prediction Error Distribution')
    ax3.set_xlabel('Error (%)')
    ax3.set_ylabel('Density')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Error Over Time
    ax4 = axes[1, 1]
    for name, pred in predictions.items():
        pred_flat = pred.flatten() if hasattr(pred, 'flatten') else pred
        error = y_test_array - pred_flat
        error_last = error[-last_252:] if len(error) > 252 else error
        ax4.plot(y_test_index_last, error_last, linewidth=1, alpha=0.7, label=name)
    ax4.axhline(y=0, color='black', linestyle='--')
    ax4.set_title('Prediction Error Over Time')
    ax4.set_xlabel('Date')
    ax4.set_ylabel('Error (%)')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Prepare predictions (ensure they're numpy arrays)
predictions = {
    'XGBoost': xgb_pred_test,
    'LightGBM': lgb_pred_test,
    'Random Forest': rf_pred_test
}

# Visualize
visualize_predictions(y_test, predictions, list(predictions.keys()))
```

```python
# Cell 14: Feature Pruning - Keep Top 30 Features
def prune_features(X_train, X_test, importance_df, top_n=30):
    """
    Keep only top N most important features
    """
    print("\n" + "="*60)
    print("FEATURE PRUNING")
    print("="*60)
    
    # Get top N features from LightGBM (or XGBoost)
    top_features = importance_df.head(top_n)['feature'].tolist()
    
    print(f"Keeping top {top_n} features:")
    for i, feat in enumerate(top_features, 1):
        print(f"  {i:2}. {feat}")
    
    # Filter data
    X_train_pruned = X_train[top_features]
    X_test_pruned = X_test[top_features]
    
    print(f"\nNew shapes - X_train: {X_train_pruned.shape}, X_test: {X_test_pruned.shape}")
    
    return X_train_pruned, X_test_pruned, top_features

# Prune to top 30 features
X_train_pruned, X_test_pruned, top_features = prune_features(X_train, X_test, lgb_importance, top_n=30)
```

```python
# Cell 15: Log Transform Target Variable (FIXED)
def transform_target(y_train, y_test):
    """
    Apply log transformation to target for better calibration
    """
    print("\n" + "="*60)
    print("LOG TRANSFORMING TARGET VARIABLE")
    print("="*60)
    
    # Convert to numpy arrays if needed
    if hasattr(y_train, 'values'):
        y_train_values = y_train.values.flatten()
        y_test_values = y_test.values.flatten()
    else:
        y_train_values = y_train.flatten()
        y_test_values = y_test.flatten()
    
    # Log transform (add small epsilon to avoid log(0))
    y_train_log = np.log(y_train_values + 1e-8)
    y_test_log = np.log(y_test_values + 1e-8)
    
    print(f"Original target stats:")
    print(f"  Mean: {np.mean(y_train_values):.4f}")
    print(f"  Std: {np.std(y_train_values):.4f}")
    print(f"  Min: {np.min(y_train_values):.4f}")
    print(f"  Max: {np.max(y_train_values):.4f}")
    
    print(f"\nLog-transformed target stats:")
    print(f"  Mean: {np.mean(y_train_log):.4f}")
    print(f"  Std: {np.std(y_train_log):.4f}")
    print(f"  Min: {np.min(y_train_log):.4f}")
    print(f"  Max: {np.max(y_train_log):.4f}")
    
    # Visualize before and after
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    axes[0].hist(y_train_values, bins=50, color='blue', alpha=0.7, edgecolor='black')
    axes[0].set_title('Original Target Distribution')
    axes[0].set_xlabel('Volatility')
    axes[0].set_ylabel('Frequency')
    
    axes[1].hist(y_train_log, bins=50, color='green', alpha=0.7, edgecolor='black')
    axes[1].set_title('Log-Transformed Target Distribution')
    axes[1].set_xlabel('log(Volatility)')
    axes[1].set_ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
    
    # Return as Series with original index
    if hasattr(y_train, 'index'):
        y_train_log_series = pd.Series(y_train_log, index=y_train.index)
        y_test_log_series = pd.Series(y_test_log, index=y_test.index)
    else:
        y_train_log_series = y_train_log
        y_test_log_series = y_test_log
    
    return y_train_log_series, y_test_log_series

# Apply log transformation
y_train_log, y_test_log = transform_target(y_train, y_test)
```

```python
# Cell 16: Add Baseline Anchor Features
def add_baseline_anchor(X_train, X_test, master_data):
    """
    Add volatility floor/ceiling features
    """
    print("\n" + "="*60)
    print("ADDING BASELINE ANCHOR FEATURES")
    print("="*60)
    
    # Get SPY returns
    spy_returns = master_data['SPY'].pct_change()
    
    # Calculate volatility floor (50-day minimum)
    vol_floor_50 = spy_returns.rolling(50).std() * np.sqrt(252)
    vol_floor_200 = spy_returns.rolling(200).std() * np.sqrt(252)
    
    # Add to feature sets
    X_train['vol_floor_50'] = vol_floor_50.reindex(X_train.index)
    X_test['vol_floor_50'] = vol_floor_50.reindex(X_test.index)
    
    X_train['vol_floor_200'] = vol_floor_200.reindex(X_train.index)
    X_test['vol_floor_200'] = vol_floor_200.reindex(X_test.index)
    
    print(f"✅ Added 2 baseline anchor features")
    print(f"New X_train shape: {X_train.shape}")
    print(f"New X_test shape: {X_test.shape}")
    
    return X_train, X_test

# Add baseline anchors
X_train_anchored, X_test_anchored = add_baseline_anchor(X_train_pruned, X_test_pruned, master_data)
```

```python
# Cell 17: Retrain with Log Target and Anchored Features
def retrain_improved_models(X_train, X_test, y_train_log, y_test_log, y_test_original):
    """
    Retrain models with all improvements
    """
    print("\n" + "="*60)
    print("RETRAINING IMPROVED MODELS")
    print("="*60)
    
    # XGBoost with Log Target
    xgb_model = xgb.XGBRegressor(
        n_estimators=200,
        learning_rate=0.03,  # Lower learning rate for stability
        max_depth=4,         # Shallower trees
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    xgb_model.fit(X_train, y_train_log)
    xgb_pred_log = xgb_model.predict(X_test)
    xgb_pred = np.exp(xgb_pred_log)  # Back-transform
    
    # LightGBM with Log Target
    lgb_model = lgb.LGBMRegressor(
        n_estimators=200,
        learning_rate=0.03,
        num_leaves=15,       # Fewer leaves for generalization
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        verbose=-1
    )
    lgb_model.fit(X_train, y_train_log)
    lgb_pred_log = lgb_model.predict(X_test)
    lgb_pred = np.exp(lgb_pred_log)  # Back-transform
    
    # Random Forest with Log Target
    rf_model = RandomForestRegressor(
        n_estimators=200,
        max_depth=6,         # Shallower trees
        min_samples_split=10,
        min_samples_leaf=5,
        random_state=42,
        n_jobs=-1
    )
    rf_model.fit(X_train, y_train_log)
    rf_pred_log = rf_model.predict(X_test)
    rf_pred = np.exp(rf_pred_log)  # Back-transform
    
    # Calculate metrics on original scale
    y_test_array = y_test_original.values.flatten()
    
    results = []
    for name, pred in [('XGBoost', xgb_pred), ('LightGBM', lgb_pred), ('Random Forest', rf_pred)]:
        rmse = np.sqrt(mean_squared_error(y_test_array, pred))
        mae = mean_absolute_error(y_test_array, pred)
        r2 = r2_score(y_test_array, pred)
        results.append({'Model': name, 'RMSE': rmse, 'MAE': mae, 'R2': r2})
        print(f"\n{name}:")
        print(f"  RMSE: {rmse:.4f}")
        print(f"  MAE: {mae:.4f}")
        print(f"  R²: {r2:.4f}")
    
    return xgb_model, lgb_model, rf_model, xgb_pred, lgb_pred, rf_pred, pd.DataFrame(results)

# Retrain with improvements
xgb_improved, lgb_improved, rf_improved, xgb_pred_imp, lgb_pred_imp, rf_pred_imp, improved_results = retrain_improved_models(
    X_train_anchored, X_test_anchored, y_train_log, y_test_log, y_test
)
```

```python
# Cell 18: Before vs After Comparison
def compare_before_after(original_comparison, improved_comparison):
    """
    Compare original vs improved model performance
    """
    print("\n" + "="*60)
    print("BEFORE vs AFTER IMPROVEMENTS")
    print("="*60)
    
    # Merge results
    merged = original_comparison.merge(improved_comparison, on='Model', suffixes=('_before', '_after'))
    
    print("\nPerformance Improvement:")
    print("-"*60)
    print(f"{'Model':<15} {'R² Before':<12} {'R² After':<12} {'Change':<10}")
    print("-"*60)
    
    for _, row in merged.iterrows():
        change = row['R2_after'] - row['R2_before']
        emoji = '✅' if change > 0 else '⚠️'
        print(f"{row['Model']:<15} {row['R2_before']:.4f}      {row['R2_after']:.4f}      {emoji} {change:+.4f}")
    
    # Visualization
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # R² Comparison
    x = np.arange(len(merged))
    width = 0.35
    axes[0].bar(x - width/2, merged['R2_before'], width, label='Before', color='red', alpha=0.6)
    axes[0].bar(x + width/2, merged['R2_after'], width, label='After', color='green', alpha=0.6)
    axes[0].set_xticks(x)
    axes[0].set_xticklabels(merged['Model'])
    axes[0].set_title('R² Score Comparison')
    axes[0].axhline(y=0, color='black', linestyle='--')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # RMSE Comparison
    axes[1].bar(x - width/2, merged['RMSE_before'], width, label='Before', color='red', alpha=0.6)
    axes[1].bar(x + width/2, merged['RMSE_after'], width, label='After', color='green', alpha=0.6)
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(merged['Model'])
    axes[1].set_title('RMSE Comparison (lower is better)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return merged

# Compare before and after
comparison_before_after = compare_before_after(comparison_df, improved_results)
```

```python
# Cell 19: Visualize Improved Predictions
def visualize_improved_predictions(y_test, predictions_original, predictions_improved):
    """
    Compare original vs improved predictions
    """
    print("\n" + "="*60)
    print("IMPROVED PREDICTION VISUALIZATION")
    print("="*60)
    
    y_test_array = y_test.values.flatten() if hasattr(y_test, 'values') else y_test.flatten()
    y_test_index = y_test.index if hasattr(y_test, 'index') else range(len(y_test_array))
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # Plot 1: Improved Predictions vs Actual
    ax1 = axes[0, 0]
    ax1.plot(y_test_index, y_test_array, 'k-', linewidth=2, label='Actual')
    for name, pred in predictions_improved.items():
        ax1.plot(y_test_index, pred, linewidth=1, alpha=0.7, label=f'{name} (Improved)')
    ax1.set_title('Improved Model Predictions vs Actual')
    ax1.set_ylabel('Volatility')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Error Comparison
    ax2 = axes[0, 1]
    for name, pred in predictions_original.items():
        error = y_test_array - pred
        ax2.hist(error, bins=30, alpha=0.3, label=f'{name} (Original)', density=True)
    for name, pred in predictions_improved.items():
        error = y_test_array - pred
        ax2.hist(error, bins=30, alpha=0.3, label=f'{name} (Improved)', density=True, linestyle='--')
    ax2.axvline(x=0, color='black', linestyle='--')
    ax2.set_title('Error Distribution: Original vs Improved')
    ax2.set_xlabel('Error')
    ax2.set_ylabel('Density')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Scatter - Improved
    ax3 = axes[1, 0]
    for name, pred in predictions_improved.items():
        ax3.scatter(y_test_array, pred, alpha=0.3, s=5, label=name)
    ax3.plot([y_test_array.min(), y_test_array.max()], [y_test_array.min(), y_test_array.max()], 
             'k--', linewidth=2, label='Perfect')
    ax3.set_title('Improved: Actual vs Predicted')
    ax3.set_xlabel('Actual')
    ax3.set_ylabel('Predicted')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Error Over Time
    ax4 = axes[1, 1]
    for name, pred in predictions_improved.items():
        error = y_test_array - pred
        ax4.plot(y_test_index, error, linewidth=1, alpha=0.7, label=name)
    ax4.axhline(y=0, color='black', linestyle='--')
    ax4.set_title('Improved: Error Over Time')
    ax4.set_xlabel('Date')
    ax4.set_ylabel('Error')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Visualize improvements
visualize_improved_predictions(
    y_test,
    {'XGBoost': xgb_pred_test, 'LightGBM': lgb_pred_test, 'Random Forest': rf_pred_test},
    {'XGBoost': xgb_pred_imp, 'LightGBM': lgb_pred_imp, 'Random Forest': rf_pred_imp}
)
```

```python
# Cell 20: Save Final Improved Models
def save_final_models(xgb_model, lgb_model, rf_model, improved_results, X_train, X_test, y_train, y_test):
    """
    Save all final models and results
    """
    import joblib
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    
    print("\n" + "="*60)
    print("SAVING FINAL IMPROVED MODELS")
    print("="*60)
    
    # Save models
    joblib.dump(xgb_model, f'final_xgb_model_{timestamp}.pkl')
    print(f"✅ Final XGBoost model saved: final_xgb_model_{timestamp}.pkl")
    
    joblib.dump(lgb_model, f'final_lgb_model_{timestamp}.pkl')
    print(f"✅ Final LightGBM model saved: final_lgb_model_{timestamp}.pkl")
    
    joblib.dump(rf_model, f'final_rf_model_{timestamp}.pkl')
    print(f"✅ Final Random Forest model saved: final_rf_model_{timestamp}.pkl")
    
    # Save scaler
    joblib.dump(scaler, f'scaler_{timestamp}.pkl')
    print(f"✅ Scaler saved: scaler_{timestamp}.pkl")
    
    # Save results
    improved_results.to_csv(f'final_ml_results_{timestamp}.csv', index=False)
    print(f"✅ Final ML results saved: final_ml_results_{timestamp}.csv")
    
    # Save feature list
    feature_list = pd.DataFrame({
        'Feature': X_train.columns.tolist(),
        'Importance': 'Selected'
    })
    feature_list.to_csv(f'final_features_{timestamp}.csv', index=False)
    print(f"✅ Final feature list saved: final_features_{timestamp}.csv")
    
    return timestamp

# Save final models
final_timestamp = save_final_models(
    xgb_improved, lgb_improved, rf_improved, 
    improved_results, X_train_anchored, X_test_anchored, 
    y_train_log, y_test
)
```

### Phase 6  portfolio construction

```python

def construct_portfolio(y_test, xgb_pred, lgb_pred, rf_pred, volatility_target=0.15):
    """
    Build portfolio using volatility forecasts
    """
    print("\n" + "="*60)
    print("PORTFOLIO CONSTRUCTION")
    print("="*60)
    
    # Use Random Forest predictions (best model)
    predicted_vol = rf_pred.flatten() if hasattr(rf_pred, 'flatten') else rf_pred
    
    # Calculate position sizes (inverse volatility)
    position_size = volatility_target / predicted_vol
    
    # Cap position sizes (risk management)
    position_size = np.clip(position_size, 0.5, 2.0)  # Min 50%, Max 200%
    
    # Create portfolio DataFrame
    portfolio = pd.DataFrame({
        'Date': y_test.index,
        'Predicted_Vol': predicted_vol,
        'Position_Size': position_size,
        'Daily_Return': np.random.normal(0, predicted_vol / np.sqrt(252), len(predicted_vol))
    })
    portfolio.set_index('Date', inplace=True)
    
    print(f"\nPortfolio Statistics:")
    print(f"  Average Position Size: {portfolio['Position_Size'].mean():.2f}")
    print(f"  Min Position Size: {portfolio['Position_Size'].min():.2f}")
    print(f"  Max Position Size: {portfolio['Position_Size'].max():.2f}")
    print(f"  Volatility Target: {volatility_target:.1%}")
    
    # Visualize position sizing
    fig, axes = plt.subplots(2, 1, figsize=(15, 8))
    
    # Plot 1: Predicted Volatility vs Position Size
    ax1 = axes[0]
    ax1.plot(portfolio.index[-252:], portfolio['Predicted_Vol'].iloc[-252:], 
             'b-', linewidth=1.5, label='Predicted Volatility')
    ax1.set_ylabel('Volatility', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.grid(True, alpha=0.3)
    
    ax1_2 = ax1.twinx()
    ax1_2.plot(portfolio.index[-252:], portfolio['Position_Size'].iloc[-252:], 
               'r-', linewidth=1.5, label='Position Size')
    ax1_2.set_ylabel('Position Size', color='red')
    ax1_2.tick_params(axis='y', labelcolor='red')
    ax1_2.axhline(y=1.0, color='black', linestyle='--', alpha=0.5)
    
    ax1.set_title('Volatility vs Position Sizing (Inverse Relationship)')
    ax1.legend(loc='upper left')
    ax1_2.legend(loc='upper right')
    
    # Plot 2: Position Size Distribution
    ax2 = axes[1]
    ax2.hist(portfolio['Position_Size'], bins=30, color='green', alpha=0.7, edgecolor='black')
    ax2.axvline(x=1.0, color='red', linestyle='--', linewidth=2, label='Normal (100%)')
    ax2.set_title('Position Size Distribution')
    ax2.set_xlabel('Position Size (Fraction of Capital)')
    ax2.set_ylabel('Frequency')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return portfolio

# Construct portfolio
portfolio = construct_portfolio(y_test, xgb_pred_imp, lgb_pred_imp, rf_pred_imp)
```

### this is code for portfolio construction
#### Positon size means total value of investment in that asset 



position_size = np.clip(position_size, 0.5, 2.0)\
What: Limits the allocation mathematically between a minimum of 0.5 ($50\%$ of normal capital allocation) and a maximum of 2.0 ($200\%$ leverage).\
Why: Risk Management Guardrail. Without this, if the model predicts a volatility near 0, the allocation formula would shoot toward infinity, instantly blowing up your account on a bad day.



#### insight 
top :=\
Inference A: Automated De-risking (The Defensive View)\
Look at the major event around 2026-03. The blue line shoots up dramatically past $0.20$. Instantly, the red line drops to its absolute minimum constraint of $0.50$.

Project Insight: This proves the system is executing successful capital preservation. When the Machine Learning model detects macro market turbulence, the portfolio automatically cuts its exposure to the minimum allowed limit. This prevents the portfolio from experiencing catastrophic drawdowns during high-risk market environments.


Inference B: Efficiency Optimization (The Aggressive View)
Look at the long valleys in the blue line between 2025-07 and 2025-11 where market volatility drops below $0.08$. During these stretches, the red line constantly hovers between $1.6$ and $2.0$.\

Project Insight: This demonstrates alpha enhancement. When the market is quiet and steady, the system realizes it is taking on less risk than its 15% target allows. It automatically leverages up (buying more position) to squeeze maximum gains out of a steady, trending environment. It acts as an offensive booster when conditions are safe.



```python
# Cell 23: Calculate Portfolio Performance Metrics
def calculate_portfolio_metrics(portfolio):
    """
    Calculate key portfolio performance metrics
    """
    print("\n" + "="*60)
    print("PORTFOLIO PERFORMANCE METRICS")
    print("="*60)
    
    # Assuming we have returns (in a real scenario, these would be from backtesting)
    returns = portfolio['Daily_Return']
    cumulative_returns = (1 + returns).cumprod()
    
    # Calculate metrics
    metrics = {
        'Total Return': cumulative_returns.iloc[-1] - 1,
        'Annualized Return': (1 + cumulative_returns.iloc[-1]) ** (252 / len(returns)) - 1,
        'Annualized Volatility': returns.std() * np.sqrt(252),
        'Sharpe Ratio': (returns.mean() / returns.std()) * np.sqrt(252) if returns.std() > 0 else 0,
        'Max Drawdown': (cumulative_returns / cumulative_returns.cummax() - 1).min(),
        'Win Rate': (returns > 0).sum() / len(returns) if len(returns) > 0 else 0,
        'Average Daily Return': returns.mean(),
        'Average Daily Volatility': portfolio['Predicted_Vol'].mean()
    }
    
    # Print metrics
    print("\nPerformance Metrics:")
    print("-"*50)
    for key, value in metrics.items():
        if 'Return' in key or 'Volatility' in key:
            print(f"{key:25}: {value:.2%}")
        else:
            print(f"{key:25}: {value:.4f}")
    
    # Visualize
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Plot 1: Cumulative Returns
    ax1 = axes[0, 0]
    ax1.plot(cumulative_returns.index, cumulative_returns, 'g-', linewidth=2)
    ax1.axhline(y=1.0, color='black', linestyle='--', alpha=0.5)
    ax1.set_title('Cumulative Portfolio Returns')
    ax1.set_ylabel('Cumulative Return')
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Drawdown
    ax2 = axes[0, 1]
    drawdown = cumulative_returns / cumulative_returns.cummax() - 1
    ax2.fill_between(drawdown.index, 0, drawdown, color='red', alpha=0.5)
    ax2.set_title('Drawdown Analysis')
    ax2.set_ylabel('Drawdown (%)')
    ax2.grid(True, alpha=0.3)
    
    # Plot 3: Rolling Sharpe Ratio
    ax3 = axes[1, 0]
    rolling_sharpe = returns.rolling(252).mean() / returns.rolling(252).std() * np.sqrt(252)
    ax3.plot(rolling_sharpe.index, rolling_sharpe, 'b-', linewidth=1.5)
    ax3.axhline(y=0, color='black', linestyle='--', alpha=0.5)
    ax3.set_title('Rolling Sharpe Ratio (1-year window)')
    ax3.set_ylabel('Sharpe Ratio')
    ax3.grid(True, alpha=0.3)
    
    # Plot 4: Return Distribution
    ax4 = axes[1, 1]
    ax4.hist(returns, bins=50, color='blue', alpha=0.7, edgecolor='black')
    ax4.axvline(x=0, color='red', linestyle='--', linewidth=2)
    ax4.set_title('Daily Return Distribution')
    ax4.set_xlabel('Daily Return')
    ax4.set_ylabel('Frequency')
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return metrics, cumulative_returns

# Calculate portfolio metrics
portfolio_metrics, cumulative_returns = calculate_portfolio_metrics(portfolio)
```
