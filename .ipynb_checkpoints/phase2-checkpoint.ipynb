```python
# Cell 3: Calculate Returns for All Assets
def calculate_returns(data):
    """
    Calculate multiple types of returns for analysis
    """
    returns_df = pd.DataFrame(index=data.index)
    
    for col in data.columns:
        # Simple returns
        returns_df[f'{col}_ret'] = data[col].pct_change()
        
        # Log returns (better for volatility modeling)
        returns_df[f'{col}_logret'] = np.log(data[col] / data[col].shift(1))
        
        # Squared returns (for volatility analysis)
        returns_df[f'{col}_sq_ret'] = returns_df[f'{col}_ret'] ** 2
        
        # Absolute returns
        returns_df[f'{col}_abs_ret'] = returns_df[f'{col}_ret'].abs()
    
    return returns_df

# Calculate returns
returns_data = calculate_returns(master_data)
print(f"✅ Calculated returns for all assets")
print(f"Shape: {returns_data.shape}")
print("\nFirst few rows:")
print(returns_data.head())
```

What's in the Returns Dataset?
From these 5 price columns, we created 20 columns:

text
Each asset (SPY, QQQ, TLT, GLD, USO) → 4 derived columns:\

1. SPY_ret        → Simple returns: (Price_t / Price_t-1) - 1
2. SPY_logret     → Log returns: ln(Price_t / Price_t-1)
3. SPY_sq_ret     → Squared returns: (SPY_ret)²
4. SPY_abs_ret    → Absolute returns: |SPY_ret|

Repeat for QQQ, TLT, GLD, USO

Total: 5 assets × 4 columns = 20 columns
This is returns data - what we analyze for volatility.

```python
# Cell 4: Descriptive Statistics of Returns
def analyze_returns(df, asset_name):
    """
    Comprehensive analysis of asset returns
    """
    ret_col = f'{asset_name}_ret'
    logret_col = f'{asset_name}_logret'
    
    if ret_col not in df.columns:
        print(f"❌ {asset_name} not found in data")
        return
    
    returns = df[ret_col].dropna()
    logreturns = df[logret_col].dropna()
    
    print(f"\n{'='*60}")
    print(f"RETURN ANALYSIS: {asset_name}")
    print('='*60)
    
    # Basic statistics
    stats_dict = {
        'Mean': returns.mean(),
        'Median': returns.median(),
        'Std Dev': returns.std(),
        'Variance': returns.var(),
        'Skewness': returns.skew(),
        'Kurtosis': returns.kurtosis(),
        'Min': returns.min(),
        'Max': returns.max(),
        'Range': returns.max() - returns.min(),
        'Sharpe Ratio (annualized)': returns.mean() / returns.std() * np.sqrt(252),
        'Positive Days (%)': (returns > 0).sum() / len(returns) * 100,
        'Negative Days (%)': (returns < 0).sum() / len(returns) * 100,
    }
    
    for key, value in stats_dict.items():
        print(f"{key:25}: {value:.6f}")
    
    # Jarque-Bera test for normality
    jb_stat, jb_pvalue = jarque_bera(returns)
    print(f"\nJarque-Bera Normality Test:")
    print(f"  Statistic: {jb_stat:.4f}")
    print(f"  p-value: {jb_pvalue:.6f}")
    if jb_pvalue < 0.05:
        print("  ❌ Returns are NOT normally distributed (reject H0)")
    else:
        print("  ✅ Returns appear normally distributed")
    
    # Shapiro-Wilk test
    if len(returns) < 5000:
        shapiro_stat, shapiro_p = shapiro(returns)
        print(f"\nShapiro-Wilk Normality Test:")
        print(f"  Statistic: {shapiro_stat:.4f}")
        print(f"  p-value: {shapiro_p:.6f}")
    
    return stats_dict

# Analyze SPY returns
spy_stats = analyze_returns(returns_data, 'SPY')

# Compare with other assets
for asset in ['QQQ', 'TLT', 'GLD', 'USO']:
    analyze_returns(returns_data, asset)
```

```python
# Cell 6: Volatility Clustering Analysis
def analyze_volatility_clustering(returns_series, asset_name, windows=[5, 10, 20, 60]):
    """
    Analyze volatility clustering using rolling standard deviation
    """
    print(f"\n{'='*60}")
    print(f"VOLATILITY CLUSTERING ANALYSIS: {asset_name}")
    print('='*60)
    
    # Calculate rolling volatility for different windows
    vol_dfs = {}
    for window in windows:
        vol = returns_series.rolling(window).std() * np.sqrt(252)
        vol_dfs[window] = vol
    
    # Create visualization
    fig, axes = plt.subplots(2, 1, figsize=(15, 8))
    
    # Plot 1: Returns with volatility bands
    ax1 = axes[0]
    ax1.plot(returns_series.index, returns_series, color='blue', alpha=0.5, linewidth=0.5)
    ax1.axhline(y=0, color='black', linestyle='-', linewidth=0.5, alpha=0.3)
    ax1.set_title(f'{asset_name} Daily Returns with Volatility Clustering')
    ax1.set_ylabel('Returns')
    ax1.grid(True, alpha=0.3)
    
    # Highlight high volatility periods
    vol_high = vol_dfs[20] > vol_dfs[20].quantile(0.75)
    ax1.fill_between(returns_series.index, -0.1, 0.1, where=vol_high, 
                     color='red', alpha=0.2, label='High Volatility Period')
    ax1.legend()
    
    # Plot 2: Rolling volatility comparison
    ax2 = axes[1]
    for window, vol in vol_dfs.items():
        ax2.plot(vol.index, vol * 100, label=f'{window}-day', alpha=0.7)
    ax2.set_title(f'{asset_name} Rolling Volatility (Annualized)')
    ax2.set_ylabel('Volatility (%)')
    ax2.set_xlabel('Date')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Summary statistics
    print("\nVolatility Summary (20-day):")
    print(f"  Mean: {vol_dfs[20].mean()*100:.2f}%")
    print(f"  Median: {vol_dfs[20].median()*100:.2f}%")
    print(f"  Std: {vol_dfs[20].std()*100:.2f}%")
    print(f"  Max: {vol_dfs[20].max()*100:.2f}%")
    print(f"  Min: {vol_dfs[20].min()*100:.2f}%")
    print(f"  Q1: {vol_dfs[20].quantile(0.25)*100:.2f}%")
    print(f"  Q3: {vol_dfs[20].quantile(0.75)*100:.2f}%")
    
    return vol_dfs

# Analyze SPY volatility clustering
spy_returns = returns_data['SPY_ret'].dropna()
spy_vols = analyze_volatility_clustering(spy_returns, 'SPY')
```

```python
# Cell 7: Autocorrelation Analysis
def analyze_autocorrelation(series, asset_name, lags=40):
    """
    Analyze autocorrelation in returns and squared returns
    """
    print(f"\n{'='*60}")
    print(f"AUTOCORRELATION ANALYSIS: {asset_name}")
    print('='*60)
    
    # Remove NaN
    series_clean = series.dropna()
    
    # Ljung-Box test for returns
    lb_returns = acorr_ljungbox(series_clean, lags=lags, return_df=True)
    
    # Ljung-Box test for squared returns
    lb_squared = acorr_ljungbox(series_clean**2, lags=lags, return_df=True)
    
    print("\nLjung-Box Test (p-values):")
    print(f"  Returns - Significant autocorrelation present: {(lb_returns['lb_pvalue'] < 0.05).any()}")
    print(f"  Squared Returns - Significant autocorrelation present: {(lb_squared['lb_pvalue'] < 0.05).any()}")
    
    # Visualize ACF and PACF
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # ACF of returns
    plot_acf(series_clean, ax=axes[0, 0], lags=lags, title=f'{asset_name} Returns - ACF')
    
    # PACF of returns
    plot_pacf(series_clean, ax=axes[0, 1], lags=lags, title=f'{asset_name} Returns - PACF')
    
    # ACF of squared returns (volatility)
    plot_acf(series_clean**2, ax=axes[1, 0], lags=lags, title=f'{asset_name} Squared Returns - ACF')
    
    # PACF of squared returns
    plot_pacf(series_clean**2, ax=axes[1, 1], lags=lags, title=f'{asset_name} Squared Returns - PACF')
    
    plt.tight_layout()
    plt.show()
    
    # Return Ljung-Box results
    return lb_returns, lb_squared

# Analyze SPY autocorrelation
spy_lb_ret, spy_lb_sq = analyze_autocorrelation(spy_returns, 'SPY')
```

```python
# Cell 8: Volatility Regime Detection (FIXED)
def detect_volatility_regimes(vol_series, n_regimes=3):
    """
    Simple volatility regime detection using percentile thresholds
    """
    print(f"\n{'='*60}")
    print("VOLATILITY REGIME DETECTION")
    print('='*60)
    
    # Remove NaN values
    vol_clean = vol_series.dropna()
    
    # Calculate thresholds
    low_thresh = vol_clean.quantile(0.33)
    high_thresh = vol_clean.quantile(0.67)
    
    # Assign regimes
    regimes = pd.Series(index=vol_clean.index, dtype='object')
    regimes[vol_clean <= low_thresh] = 'Low'
    regimes[(vol_clean > low_thresh) & (vol_clean <= high_thresh)] = 'Medium'
    regimes[vol_clean > high_thresh] = 'High'
    
    # Remove any NaN values
    regimes = regimes.dropna()
    
    print(f"Regime Thresholds:")
    print(f"  Low: < {low_thresh*100:.2f}%")
    print(f"  Medium: {low_thresh*100:.2f}% - {high_thresh*100:.2f}%")
    print(f"  High: > {high_thresh*100:.2f}%")
    print(f"\nRegime Distribution:")
    print(f"  Low: {(regimes == 'Low').sum()} days ({((regimes == 'Low').sum()/len(regimes))*100:.1f}%)")
    print(f"  Medium: {(regimes == 'Medium').sum()} days ({((regimes == 'Medium').sum()/len(regimes))*100:.1f}%)")
    print(f"  High: {(regimes == 'High').sum()} days ({((regimes == 'High').sum()/len(regimes))*100:.1f}%)")
    
    # Visualize regimes
    fig, axes = plt.subplots(2, 1, figsize=(15, 8))
    
    # Plot 1: Volatility with regimes
    ax1 = axes[0]
    ax1.plot(vol_clean.index, vol_clean * 100, color='blue', linewidth=1)
    ax1.axhline(y=low_thresh*100, color='green', linestyle='--', linewidth=1, label=f'Low ({(low_thresh*100):.1f}%)')
    ax1.axhline(y=high_thresh*100, color='red', linestyle='--', linewidth=1, label=f'High ({(high_thresh*100):.1f}%)')
    
    # Color background by regime - FIXED: Use clean data
    ax1.fill_between(vol_clean.index, 0, vol_clean.max()*100, 
                     where=vol_clean <= low_thresh, 
                     color='green', alpha=0.2, label='Low Vol Regime')
    ax1.fill_between(vol_clean.index, 0, vol_clean.max()*100, 
                     where=(vol_clean > low_thresh) & (vol_clean <= high_thresh), 
                     color='yellow', alpha=0.2, label='Medium Vol Regime')
    ax1.fill_between(vol_clean.index, 0, vol_clean.max()*100, 
                     where=vol_clean > high_thresh, 
                     color='red', alpha=0.2, label='High Vol Regime')
    
    ax1.set_title('Volatility Regimes (20-day Rolling Volatility)')
    ax1.set_ylabel('Volatility (%)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Regime timeline - FIXED: Handle NaN values
    ax2 = axes[1]
    regime_colors = {'Low': 'green', 'Medium': 'yellow', 'High': 'red'}
    
    # Filter out any remaining NaN values
    valid_indices = ~regimes.isna()
    regimes_clean = regimes[valid_indices]
    
    ax2.scatter(regimes_clean.index, [0]*len(regimes_clean), 
                c=[regime_colors[r] for r in regimes_clean], 
                s=5, alpha=0.6)
    ax2.set_title('Volatility Regime Timeline')
    ax2.set_yticks([])
    ax2.set_xlabel('Date')
    ax2.grid(True, alpha=0.3)
    
    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [Patch(facecolor='green', label='Low Vol'),
                       Patch(facecolor='yellow', label='Medium Vol'),
                       Patch(facecolor='red', label='High Vol')]
    ax2.legend(handles=legend_elements, loc='upper right')
    
    plt.tight_layout()
    plt.show()
    
    return regimes_clean, low_thresh, high_thresh

# Detect regimes for SPY
spy_vol = spy_vols[20]  # Use 20-day volatility
spy_regimes, low_thresh, high_thresh = detect_volatility_regimes(spy_vol)
```

### If i say there is volatility of ...% today , computer does not know what to do with numbers.
regime means specific market env here high low medium
Low Regime (< 10.36%): The "Sunny" market. Investors are highly confident. Prices drift upward smoothly. Action: The trading model can safely take on high leverage and buy lots of stocks.\
Medium Regime (10.36% - 16.41%): The "Cloudy" market. Typical everyday trading. Standard economic ups and downs. Action: The trading model behaves normally, holding a balanced mix of assets.\
High Regime (> 16.41%): The "Stormy" market. Panic, crashes, and chaos.\
Action: The trading model hits the panic button, slashes stock investments, and hides capital in safe-haven assets (Gold/Bonds).

We developed a non-linear Volatility Regime Detection model using historical percentile thresholds (33rd and 67th quantiles). This transforms raw continuous volatility numbers into discrete, operational environments: Low (<10.36%), Medium (10.36%-16.41%), and High (>16.41%). The primary significance of this step is to break away from the traditional financial assumption that markets behave identically at all times. The timeline visualization explicitly shows that market states are persistent; when a high-risk regime (red band) triggers, it clusters and remains active for a prolonged structural era. This regime categorization serves as the ultimate switch-board engine for our final asset allocation model, allowing the algorithm to dynamically shift portfolio weights as the market shifts from calm to crisis states.

```python
# Cell 9: Analyze Regime-Specific Statistics (FIXED)
def analyze_regime_statistics(returns, regimes):
    """
    Calculate statistics for each volatility regime
    """
    print("\n" + "="*60)
    print("REGIME-SPECIFIC STATISTICS")
    print("="*60)
    
    # Ensure we have clean data
    returns_clean = returns.dropna()
    regimes_clean = regimes.dropna()
    
    # Align indices
    common_idx = returns_clean.index.intersection(regimes_clean.index)
    returns_aligned = returns_clean.loc[common_idx]
    regimes_aligned = regimes_clean.loc[common_idx]
    
    for regime in ['Low', 'Medium', 'High']:
        mask = regimes_aligned == regime
        regime_returns = returns_aligned.loc[mask]
        
        if len(regime_returns) > 0:
            print(f"\n{regime} Volatility Regime:")
            print(f"  Count: {len(regime_returns)} days")
            print(f"  Mean Return: {regime_returns.mean():.6f}")
            print(f"  Std Return: {regime_returns.std():.6f}")
            print(f"  Sharpe Ratio: {regime_returns.mean()/regime_returns.std():.4f}")
            print(f"  Win Rate: {(regime_returns > 0).sum()/len(regime_returns)*100:.1f}%")
            print(f"  Average Win: {regime_returns[regime_returns > 0].mean():.6f}")
            print(f"  Average Loss: {regime_returns[regime_returns < 0].mean():.6f}")
            print(f"  Max Daily Return: {regime_returns.max():.6f}")
            print(f"  Min Daily Return: {regime_returns.min():.6f}")

# Analyze regime statistics
spy_returns_clean = returns_data['SPY_ret'].dropna()
analyze_regime_statistics(spy_returns_clean, spy_regimes)
```

```python
# Cell 10: Volatility Cone Analysis
def analyze_volatility_cone(returns, asset_name, windows=[5, 10, 20, 60, 120]):
    """
    Create volatility cone showing expected volatility ranges
    """
    print(f"\n{'='*60}")
    print(f"VOLATILITY CONE ANALYSIS: {asset_name}")
    print('='*60)
    
    # Calculate volatility for different windows
    vol_cone = {}
    for window in windows:
        vol = returns.rolling(window).std() * np.sqrt(252)
        vol_cone[window] = vol
    
    # Create visualization
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Calculate percentiles
    percentiles = [5, 25, 50, 75, 95]
    
    for idx, window in enumerate(windows):
        vol_data = vol_cone[window].dropna()
        if len(vol_data) > 0:
            percentiles_data = vol_data.quantile([p/100 for p in percentiles])
            
            # Plot median
            ax.plot([window] * len(percentiles_data), percentiles_data.values, 
                    'o-', color='darkblue', markersize=8)
            
            # Add shaded area for percentiles
            ax.fill_betweenx([percentiles_data.iloc[0], percentiles_data.iloc[-1]], 
                            window-0.5, window+0.5, 
                            color='lightblue', alpha=0.3)
            ax.fill_betweenx([percentiles_data.iloc[1], percentiles_data.iloc[-2]], 
                            window-0.5, window+0.5, 
                            color='steelblue', alpha=0.3)
            ax.fill_betweenx([percentiles_data.iloc[2]-0.1, percentiles_data.iloc[2]+0.1], 
                            window-0.5, window+0.5, 
                            color='red', alpha=0.5)
    
    ax.set_xlabel('Lookback Window (Days)')
    ax.set_ylabel('Annualized Volatility')
    ax.set_title(f'{asset_name} Volatility Cone')
    ax.set_xticks(windows)
    ax.grid(True, alpha=0.3)
    
    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='lightblue', alpha=0.7, label='5th-95th Percentile'),
        Patch(facecolor='steelblue', alpha=0.7, label='25th-75th Percentile'),
        Patch(facecolor='red', alpha=0.7, label='Median')
    ]
    ax.legend(handles=legend_elements, loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary
    print("\nSummary Statistics (Volatility Cone):")
    for window in windows:
        vol_data = vol_cone[window].dropna()
        if len(vol_data) > 0:
            print(f"\n  {window}-day window:")
            print(f"    Mean: {vol_data.mean()*100:.2f}%")
            print(f"    Median: {vol_data.median()*100:.2f}%")
            print(f"    Std: {vol_data.std()*100:.2f}%")
            print(f"    Range: {vol_data.min()*100:.2f}% - {vol_data.max()*100:.2f}%")
    
    return vol_cone

# Analyze volatility cone for SPY
vol_cone = analyze_volatility_cone(spy_returns, 'SPY')
```

Horizontal axis (x-axis) : length of time used to measure the market price swing
Y axis :- annualized volatility 



```python
# Cell 11: Cross-Asset Volatility Analysis (FIXED)
def analyze_cross_asset_volatility(returns_data):
    """
    Compare volatility across all assets
    """
    print("\n" + "="*60)
    print("CROSS-ASSET VOLATILITY ANALYSIS")
    print("="*60)
    
    # Calculate 20-day rolling volatility for all assets
    vol_data = pd.DataFrame()
    for col in returns_data.columns:
        if '_ret' in col and not '_logret' in col and not '_sq_ret' in col and not '_abs_ret' in col:
            asset = col.replace('_ret', '')
            vol_data[asset] = returns_data[col].rolling(20).std() * np.sqrt(252)
    
    # Remove any columns with all NaN
    vol_data = vol_data.dropna(axis=1, how='all')
    
    # Summary statistics - FIXED: Handle dynamic number of columns
    print("\n20-Day Rolling Volatility Summary (Annualized):")
    stats_df = vol_data.describe().T * 100
    
    if len(stats_df) > 0:
        # Create column names dynamically based on what describe() returns
        col_names = []
        for col in stats_df.columns:
            if col == 'count':
                col_names.append('Count')
            elif col == 'mean':
                col_names.append('Mean (%)')
            elif col == 'std':
                col_names.append('Std (%)')
            elif col == 'min':
                col_names.append('Min (%)')
            elif col == '25%':
                col_names.append('Q1 (%)')
            elif col == '50%':
                col_names.append('Median (%)')
            elif col == '75%':
                col_names.append('Q3 (%)')
            elif col == 'max':
                col_names.append('Max (%)')
        stats_df.columns = col_names
        print(stats_df.round(2))
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Box plot
    ax1 = axes[0, 0]
    vol_data.boxplot(ax=ax1)
    ax1.set_title('Volatility Distribution by Asset')
    ax1.set_ylabel('Annualized Volatility')
    ax1.grid(True, alpha=0.3)
    
    # Time series of all assets
    ax2 = axes[0, 1]
    for col in vol_data.columns:
        if not vol_data[col].isna().all():
            ax2.plot(vol_data.index, vol_data[col] * 100, label=col, alpha=0.7, linewidth=1)
    ax2.set_title('Volatility Over Time - All Assets')
    ax2.set_ylabel('Volatility (%)')
    ax2.legend(loc='upper right')
    ax2.grid(True, alpha=0.3)
    
    # Correlation matrix
    ax3 = axes[1, 0]
    corr = vol_data.corr()
    if not corr.empty:
        sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', ax=ax3)
        ax3.set_title('Volatility Correlation Matrix')
    
    # Scatter plot of SPY vs QQQ volatility if both exist
    ax4 = axes[1, 1]
    if 'SPY' in vol_data.columns and 'QQQ' in vol_data.columns:
        ax4.scatter(vol_data['SPY'] * 100, vol_data['QQQ'] * 100, alpha=0.3, s=10)
        ax4.plot([0, 100], [0, 100], 'r--', alpha=0.5, label='1:1 Line')
        ax4.set_xlabel('SPY Volatility (%)')
        ax4.set_ylabel('QQQ Volatility (%)')
        ax4.set_title('SPY vs QQQ Volatility Relationship')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return vol_data

# Analyze cross-asset volatility
cross_vol = analyze_cross_asset_volatility(returns_data)
```

```python
# Cell 13: Volatility Persistence Analysis (FIXED)
def analyze_volatility_persistence(vol_series, lags=50):
    """
    Analyze how long volatility persists
    """
    print("\n" + "="*60)
    print("VOLATILITY PERSISTENCE ANALYSIS")
    print("="*60)
    
    # Remove NaN values
    vol_clean = vol_series.dropna()
    
    if len(vol_clean) < 10:
        print("⚠️ Not enough data points for analysis")
        return None, None
    
    # Adjust lags based on data length
    max_lags = min(lags, len(vol_clean) // 2 - 1)
    if max_lags < 5:
        print(f"⚠️ Not enough data for meaningful autocorrelation analysis")
        return None, None
    
    print(f"Using {max_lags} lags for analysis")
    
    # Calculate autocorrelation of volatility
    acf_values = acf(vol_clean, nlags=max_lags)
    
    # Find where autocorrelation drops below 0.5
    half_life = np.where(acf_values < 0.5)[0]
    if len(half_life) > 0:
        half_life_days = half_life[0]
    else:
        half_life_days = len(acf_values)
    
    print(f"\nVolatility Autocorrelation:")
    if len(acf_values) > 1:
        print(f"  Lag 1: {acf_values[1]:.4f}")
    if len(acf_values) > 5:
        print(f"  Lag 5: {acf_values[5]:.4f}")
    if len(acf_values) > 10:
        print(f"  Lag 10: {acf_values[10]:.4f}")
    if len(acf_values) > 20:
        print(f"  Lag 20: {acf_values[20]:.4f}")
    if len(acf_values) > 50:
        print(f"  Lag 50: {acf_values[50]:.4f}")
    print(f"\nApproximate Half-life: {half_life_days} days")
    
    # Visualization
    fig, axes = plt.subplots(2, 1, figsize=(15, 8))
    
    # Plot 1: Autocorrelation function
    ax1 = axes[0]
    ax1.bar(range(max_lags+1), acf_values, color='steelblue', alpha=0.7)
    ax1.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
    ax1.axhline(y=0.5, color='red', linestyle='--', linewidth=1, label='Half-life threshold')
    ax1.axvline(x=half_life_days, color='green', linestyle='--', linewidth=1, 
                label=f'Half-life: {half_life_days} days')
    ax1.set_title('Volatility Autocorrelation')
    ax1.set_xlabel('Lag')
    ax1.set_ylabel('Autocorrelation')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Plot 2: Volatility over time with persistence markers
    ax2 = axes[1]
    ax2.plot(vol_clean.index, vol_clean * 100, color='blue', linewidth=1)
    ax2.set_title('Volatility with Persistence Indicators')
    ax2.set_ylabel('Volatility (%)')
    ax2.grid(True, alpha=0.3)
    
    # Add markers for significant volatility events
    high_vol_threshold = vol_clean.quantile(0.75)
    high_vol_mask = vol_clean > high_vol_threshold
    if high_vol_mask.any():
        ax2.scatter(vol_clean[high_vol_mask].index, 
                    vol_clean[high_vol_mask] * 100, 
                    color='red', s=20, alpha=0.5, label='High Vol (>75th percentile)')
        ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    return acf_values, half_life_days

# Analyze volatility persistence
acf_values, half_life = analyze_volatility_persistence(spy_vol)
```

### For how many days does that memory remain active before fading away completely?
this graph shows that

The Half-life is the absolute most important number for your trading strategy. It measures the exact number of days it takes for a market panic to lose 50% of its initial energy.

What this tells your trading model:\
If a catastrophic black swan crash shocks the stock market today, your trading model can mathematically expect that the resulting wave of high-risk turbulence will take 31 trading days (roughly 1.5 calendar months) to settle halfway back down to a calm baseline.


We utilized the 75th percentile threshold as our statistical boundary line for regime classification. By isolating the top 25% highest volatility observations in our 20-year dataset, we remove human bias from risk definition. This percentile filter allows our system to dynamically map exactly when a standard market fluctuation crosses over into a historic, high-risk persistence state (represented by the clusters of red dots on our timeline).

Step 1: Calculate the Base Correlation (Lag 1)First, the code calculates the correlation between yesterday’s volatility and today’s volatility. This is your starting point (Lag 1), which came out to 0.9933. This means the initial memory of market risk is almost at 100%.\
Step 2: Establish the "Half-Life Target"The term "Half-Life" means we want to find the exact point where the data's predictive power drops to half of its initial value.The computer takes your starting Lag 1 value (0.9933) and divides it by 2.This establishes your target threshold: \(0.9933 \div 2 = \mathbf{0.4966}\) (roughly 0.50).This is exactly why you see the horizontal red dashed line drawn right at the 0.5 mark on your top chart.\
Step 3: Tracking the Daily Decay (Exponential Decay)As you look further out in time (Lag 2, Lag 3, Lag 4...), the days lose connection to the original shock. The correlation decays exponentially. The computer looks down your data table to see when the score drops below that 0.4966 target line:\
At Lag 20, the score is still 0.6464 (higher than 0.50).\
At Lag 30, the score is drifting right around 0.5100 (almost there).\
At Lag 31, the calculated autocorrelation drops to roughly 0.4950.Because \
Lag 31 is the exact day where the correlation value slips underneath the red target line,\
the computer draws the vertical green line there and declares the half-life to be 31 days.
