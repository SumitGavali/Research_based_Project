```python
# Cell 2: Create FastAPI App Code (FIXED)
# This will be saved as 'api/app.py'

fastapi_code = '''
# ============================================
# FASTAPI APPLICATION FOR VOLATILITY FORECASTING
# ============================================

from fastapi import FastAPI, HTTPException
from fastapi.responses import JSONResponse
from pydantic import BaseModel
import pandas as pd
import numpy as np
import joblib
from datetime import datetime, timedelta
import yfinance as yf
from typing import Optional, List
import uvicorn

# Initialize FastAPI app
app = FastAPI(
    title="Volatility Forecasting API",
    description="Predict market volatility using GARCH and Machine Learning models",
    version="1.0.0"
)

# Request models
class PredictionRequest(BaseModel):
    ticker: str = "SPY"
    horizon: int = 5
    target_vol: float = 0.15

class BatchPredictionRequest(BaseModel):
    tickers: List[str] = ["SPY", "QQQ", "TLT"]
    horizon: int = 5

# Response models
class PredictionResponse(BaseModel):
    ticker: str
    date: str
    predicted_volatility: float
    position_size: float
    confidence_interval: dict

class HealthResponse(BaseModel):
    status: str
    timestamp: str
    model_loaded: bool

# Global variables
models = {}
scaler = None
feature_names = None

# ============================================
# LOAD MODELS
# ============================================

def load_models():
    """Load all trained models"""
    global models, scaler, feature_names
    
    try:
        # Load Random Forest model (best performer)
        models['rf'] = joblib.load('models/final_rf_model.pkl')
        print("✅ Random Forest model loaded")
        
        # Load scaler
        scaler = joblib.load('models/scaler.pkl')
        print("✅ Scaler loaded")
        
        # Load feature names
        feature_names = models['rf'].feature_names_in_
        print(f"✅ Features loaded: {len(feature_names)} features")
        
        return True
    except Exception as e:
        print(f"❌ Error loading models: {e}")
        return False

# Load models on startup
load_models()

# ============================================
# HELPER FUNCTIONS
# ============================================

def get_market_data(ticker, days=100):
    """Fetch market data from Yahoo Finance"""
    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=days)
        
        data = yf.download(ticker, start=start_date, end=end_date, progress=False)
        return data
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None

def calculate_features(data):
    """Calculate features for prediction"""
    # This is a simplified version - in production, use the full feature pipeline
    features = {}
    
    # Returns
    returns = data['Close'].pct_change()
    
    # Lag features
    for lag in [1, 2, 3, 5, 10]:
        features[f'SPY_lag{lag}'] = returns.shift(lag).iloc[-1]
    
    # Rolling volatility
    for window in [5, 10, 20]:
        features[f'SPY_roll_std_{window}'] = returns.rolling(window).std().iloc[-1]
        features[f'SPY_hist_vol_{window}'] = returns.rolling(window).std().iloc[-1] * np.sqrt(252)
    
    # Volatility floor
    features['vol_floor_50'] = returns.rolling(50).std().iloc[-1] * np.sqrt(252)
    features['vol_floor_200'] = returns.rolling(200).std().iloc[-1] * np.sqrt(252)
    
    return features

def create_feature_vector(data, feature_names):
    """Create feature vector for prediction"""
    features_dict = calculate_features(data)
    
    # Create feature vector in correct order
    feature_vector = []
    for name in feature_names:
        if name in features_dict:
            feature_vector.append(features_dict[name])
        else:
            feature_vector.append(0)  # Default for missing features
    
    return np.array(feature_vector).reshape(1, -1)

# ============================================
# API ENDPOINTS
# ============================================

@app.get("/")
async def root():
    """Root endpoint"""
    return {
        "message": "Volatility Forecasting API",
        "version": "1.0.0",
        "status": "running",
        "docs": "/docs"
    }

@app.get("/health")
async def health():
    """Health check endpoint"""
    return HealthResponse(
        status="healthy",
        timestamp=datetime.now().isoformat(),
        model_loaded=bool(models.get('rf'))
    )

@app.post("/predict", response_model=PredictionResponse)
async def predict(request: PredictionRequest):
    """Predict future volatility for a given ticker"""
    try:
        # Fetch data
        data = get_market_data(request.ticker, days=100)
        if data is None or len(data) < 50:
            raise HTTPException(status_code=404, detail="Insufficient data for ticker")
        
        # Create feature vector
        feature_vector = create_feature_vector(data, feature_names)
        if scaler:
            feature_vector = scaler.transform(feature_vector)
        
        # Make prediction
        prediction_log = models['rf'].predict(feature_vector)[0]
        prediction = np.exp(prediction_log)  # Convert from log space
        
        # Calculate position size
        position_size = request.target_vol / prediction
        position_size = np.clip(position_size, 0.3, 2.0)
        
        return PredictionResponse(
            ticker=request.ticker,
            date=datetime.now().strftime("%Y-%m-%d"),
            predicted_volatility=float(prediction),
            position_size=float(position_size),
            confidence_interval={
                "lower": float(prediction * 0.8),
                "upper": float(prediction * 1.2)
            }
        )
    
    except HTTPException:
        raise
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

@app.post("/predict/batch")
async def batch_predict(request: BatchPredictionRequest):
    """Predict volatility for multiple tickers"""
    results = []
    
    for ticker in request.tickers:
        try:
            pred_request = PredictionRequest(ticker=ticker, horizon=request.horizon)
            result = await predict(pred_request)
            results.append(result.dict())
        except Exception as e:
            results.append({
                "ticker": ticker,
                "error": str(e)
            })
    
    return {
        "timestamp": datetime.now().isoformat(),
        "results": results
    }

@app.get("/models")
async def list_models():
    """List available models"""
    return {
        "available_models": list(models.keys()),
        "feature_count": len(feature_names) if feature_names else 0
    }

# ============================================
# RUN SERVER
# ============================================

if __name__ == "__main__":
    uvicorn.run(app, host="0.0.0.0", port=8000)
'''

# Save the FastAPI code
with open('api_app.py', 'w') as f:
    f.write(fastapi_code)
print("✅ FastAPI application created: api_app.py")
```
